#### patient( ) - 

```python
patient='L81'
```

```python
# Pull data for patient 
p=sql.read_sql("select Grammy_Table.Sample_ID,Tax_Table.%s,SUM(Grammy_Table.Normalized_Abundance) "
                 "from Tax_Table "
                 "INNER JOIN Grammy_Table "
                 "ON Tax_Table.tax_id=Grammy_Table.TaxID "
                 "WHERE Grammy_Table.Patient_ID = '%s' "
                 "GROUP BY Grammy_Table.Sample_ID,Tax_Table.%s "%(level,patient,level),engine)

p=p[~pd.isnull(p[level])]
p.set_index(['sample_id',level],inplace=True)
p=p.unstack()['sum'].fillna(0).transpose() 
    
# Sort patient data by date
s=sql.read_sql("select Sample_Table.Sample_ID,Sample_Table.DaysPostTransplant "
                 "from Sample_Table "
                 "WHERE Sample_Table.Patient_ID = '%s' "%patient,engine)

s.set_index('sample_id',inplace=True)
s['daysposttransplant']=s['daysposttransplant'].astype(float)
sam_ix=s.loc[p.columns]
sam_ix.sort('daysposttransplant',inplace=True)
p=p[sam_ix.index]

# Sort patient data by abundance
tax_sum=p.sum(axis=1)
tax_sum.sort(ascending=False)
p=p.loc[tax_sum.index]

# Pull data for cohort, which is used to compute percentiles.
if cohort == 'All':
    q=all_studies
else:
    q=" '%s' "%cohort
cohort_dat=sql.read_sql("select Grammy_Table.Sample_ID,Tax_Table.%s,SUM(Grammy_Table.Normalized_Abundance) "
                 "from Tax_Table "
                 "INNER JOIN Grammy_Table "
                 "ON Tax_Table.tax_id=Grammy_Table.TaxID "
                 "WHERE Grammy_Table.Study in (%s) "
                 "GROUP BY Grammy_Table.Sample_ID,Tax_Table.%s "%(level,q,level),engine)
cohort_dat=cohort_dat[~pd.isnull(cohort_dat[level])] # Only include abundance of bugs at specified tax level.
cohort_dat.set_index(['sample_id',level],inplace=True)
cohort_dat=cohort_dat.unstack()['sum'].fillna(0).transpose()

# Compute and store percentile data
stor_percen=pd.DataFrame(index=p.index,columns=p.columns)
for tax in p.index:
    tax_meas_ge=np.array(p.loc[tax])  # TaxID measured across samples for patient.              
    cohort_data_for_taxID=np.array(cohort_dat.loc[tax]) # TaxID measured across full cohort.
    tax_meas_percen=[stats.percentileofscore(cohort_data_for_taxID, j, kind='strict') for j in tax_meas_ge]
    stor_percen.loc[tax,:]=np.round(tax_meas_percen,decimals=2)
    
# Assign bug name and link to patient-infection page
stor_percen['name']=stor_percen.index
stor_percen['name']=stor_percen['name'].astype(int).astype(str)
stor_percen['name']=stor_percen['name'].apply(lambda x: "<a href='%s/?level=%s'>%s</a>" % (x,level,getName(x)))
stor_percen=stor_percen[['name']+list(sam_ix.index)]

# Assign column name and link to sample page
sam_names=pd.Series(stor_percen.columns)
stor_percen.columns=sam_names.apply(lambda x: "<a href='sample/%s/?level=%s'>%s</a>" % (x,level,x))
stor_percen.reset_index(inplace=True,drop=True)
stor_percen.head(3)
```

```python
# Cohort query
if cohort == 'All':
    q=all_studies
else:
    q=" '%s' "%cohort

# Pull data for cohort, which is used to compute percentiles.
cohort_dat=sql.read_sql("select Grammy_Table.Sample_ID,Tax_Table.%s,SUM(Grammy_Table.Normalized_Abundance) "
                 "from Tax_Table "
                 "INNER JOIN Grammy_Table "
                 "ON Tax_Table.tax_id=Grammy_Table.TaxID "
                 "WHERE Grammy_Table.Study in (%s) "
                 "GROUP BY Grammy_Table.Sample_ID,Tax_Table.%s "%(level,q,level),engine)

cohort_dat=cohort_dat[~pd.isnull(cohort_dat[level])] # Only include abundance of bugs at specified tax level.
cohort_dat.set_index(['sample_id',level],inplace=True)
cohort_dat=cohort_dat.unstack()['sum'].fillna(0).transpose()

# Pull data for patient 
sam_dat=sql.read_sql("select Tax_Table.%s,SUM(Grammy_Table.Normalized_Abundance) "
                 "from Tax_Table "
                 "INNER JOIN Grammy_Table "
                 "ON Tax_Table.tax_id=Grammy_Table.TaxID "
                 "WHERE Grammy_Table.Sample_ID = '%s' "
                 "GROUP BY Grammy_Table.Sample_ID,Tax_Table.%s "%(level,sample,level),engine)

# Genome equivalents for each taxID
sam_dat=sam_dat[~pd.isnull(sam_dat[level])] # Only include abundance of bugs at specified tax level.
sam_dat.set_index([level],inplace=True)
sam_dat.sort('sum',inplace=True,ascending=False)
sam_dat.columns=['gpm']

# Percentile data for each taxID
tax_meas_percen=[stats.percentileofscore(cohort_dat.loc[tax], sam_dat.loc[tax,'gpm'], kind='strict') for tax in sam_dat.index]
sam_dat['percentile']=tax_meas_percen

# Pull blast for patient 
sam_blast=sql.read_sql("select Tax_Table.%s,SUM(Grammy_Table.grammyHits) "
                 "from Tax_Table "
                 "INNER JOIN Grammy_Table "
                 "ON Tax_Table.tax_id=Grammy_Table.TaxID "
                 "WHERE Grammy_Table.Sample_ID = '%s' "
                 "GROUP BY Grammy_Table.Sample_ID,Tax_Table.%s "%(level,sample,level),engine)

# Blast for each taxID
sam_blast=sam_blast[~pd.isnull(sam_blast[level])] # Only include abundance of bugs at specified tax level.
sam_blast.set_index([level],inplace=True)
sam_blast.columns=['blast']

# Merge blast with the GPM, Percentile data
with_blast=pd.merge(sam_dat,sam_blast,left_index=True,right_index=True)

# Final details
scale=1000000
with_blast['gpm']=with_blast['gpm']*scale
with_blast['name']=with_blast.index
with_blast['name']=with_blast['name'].apply(lambda x: "<a href='/browser/%s/%s/%s/?level=%s'>%s</a>" % (cohort,patient,x,level,getName(x)))
with_blast=with_blast[['name','gpm','percentile','blast']]
with_blast.reset_index(inplace=True,drop=True)
with_blast.head(5)
```

#### coverage() -

```python
patient='L81'
taxID='1743' # 1743, 48736

# All taxIDs identified under the parent taxID
tax_table_taxID=sql.read_sql("select tax_id,tax_name,%s "
                        "from Tax_Table " 
                        "WHERE Tax_Table.%s = '%s' "%(level,level,taxID),engine)
tmp=tax_table[tax_table[level]==int(taxID)] # Text in SQL, but float / int in Pandas. 
tree=list(tmp.index)
t=[str(t) for t in tree]
t_query=','.join(t)

# BLAST hits in patient that match any child taxIDs
blast_table_sql=sql.read_sql("select * "
                 "from Blast_Table "
                 "WHERE (Blast_Table.Patient_ID = '%s' AND Blast_Table.TaxID in (%s)) " %(patient,t_query),engine)

# Lengths per GI
gi_dict=sql.read_sql("select * from GI_length_table ",engine)
gi_dict.set_index('gi',inplace=True)

# TaxIDs present with corresponding GIs
tax_gi=blast_table_sql[['gi','taxid']]
tax_to_gi=tax_gi.groupby('taxid').gi.apply(lambda x:list(set(x)))

# BLAST hits table to slice from
blast_table_sql=blast_table_sql.set_index(['gi','taxid'])

# Store read start across GIs under each taxID
taxID_store={}
size_all_gis={}
for tax_id_toquery in tax_to_gi.index:
    # GIs under the taxID
    gis_to_query=list(tax_to_gi.loc[tax_id_toquery])
    gi_sizes=gi_dict.loc[gis_to_query]
    gi_sizes.sort('length',inplace=True,ascending=True)
    # Stor GI
    stor=np.array([])
    for i,gi in enumerate(gi_sizes.index):
        pulled=blast_table_sql.loc[gi].loc[tax_id_toquery] # Blast data for GI
        if i == 0:
            gis_to_process=np.array(pulled['sstart']) # Record start position for reads hitting first GI
        else:
            gis_to_process=np.array(pulled['sstart'])+gi_sizes.loc[gi_sizes.index[i-1],'length']  # Add prior GI size to recorded start
        stor=np.append(stor,gis_to_process)
    taxID_store[tax_id_toquery]=stor
    size_all_gis[tax_id_toquery]=gi_sizes.sum(axis=0)[0]
    
# For each taxID, compute histogram of reads across GIs.
nbins=50
binNames=["%s"%i for i in list(range(nbins)) if i>0]
df=pd.DataFrame(columns=binNames)
for k in taxID_store.keys():    
    bins=np.linspace(1,np.max(taxID_store[k]),nbins)
    bins=np.linspace(1,size_all_gis[k],nbins)
    hist_vals=np.histogram(taxID_store[k],bins=bins)
    df.loc[k]=hist_vals[0]
sums=df.sum(axis=1)
sums.sort(ascending=False)
df=df.loc[sums.index]

fig,(ax1,ax2) = plt.subplots(2,1)

df_vec=df.sum(axis=0)
ind=np.arange(df_vec.shape[0])
ax1.bar(ind,np.array(df_vec),color=plot_color,linewidth=1.5)
ax1.set_xticklabels([])
ax1.set_xlim([ind[0],ind[-1]]) 
plt.setp(ax1.get_yticklabels(), rotation='horizontal', fontsize=8)
ax1.set_ylabel('Sum reads per position across taxa', fontsize = 10)
ax1.set_xlabel('Genomic bin', fontsize = 10)
ax1.set_title('Histogram of reads per genomic for each tax under: %s'%getName(taxID),fontsize = 10)

df_norm=df.div(df.sum(axis=1),axis=0)
df_norm.dropna(axis=0,inplace=True)
labels=[getName(tax) for tax in df_norm.index]
df_norm.index=labels
ax2.pcolormesh(np.array(df_norm),cmap='Reds')
ax2.set_yticks(np.arange(df_norm.shape[0])+0.5)
ax2.set_yticklabels(labels)
plt.setp(ax2.get_xticklabels(), rotation='vertical', fontsize=6)
ax2.set_xticklabels([])
ax2.set_xlim([ind[0],ind[-1]]) 
plt.setp(ax2.get_yticklabels(), rotation='horizontal', fontsize=6)

plt.tight_layout()
```
