```python
def makeHist(bs_dat,type_dat):

    # Compute histogram of the data to evaluate discretization
    all_data_unstack=bs_dat.unstack()
    # Basic statistics
    print "Shape of all bugs across all sites: %s"%all_data_unstack.shape[0]
    zeros=all_data_unstack[all_data_unstack==0]
    print "Shape of zeros data: %s"%zeros.shape[0]
    nonzeros=np.log10(all_data_unstack[all_data_unstack!=0])
    print "Shape of nonzero data: %s"%nonzeros.shape[0]
    
    plt.hist(nonzeros, bins=np.linspace(np.min(nonzeros),np.max(nonzeros),10),color='k')
    plt.xticks(fontsize=5)
    plt.yticks(fontsize=5)
    plt.xlabel('log10(mean fractional abundance)',fontsize=6)
    plt.ylabel('# microbe measurments',fontsize=6)
    plt.title('%s: Histogram of non-zero mean fractional abundances \n total %s and %s zero values.'%(type_dat,all_data_unstack.shape[0],zeros.shape[0]),fontsize=6)
    
plt.subplot(2,2,1)
makeHist(mean_data_16s,'16s')

plt.subplot(2,2,2)
makeHist(mean_data_hmp,'HMP')

plt.tight_layout()
```

```python
def discretize(bs_dat,type_dat):

    # Copy to compute specificity
    bs_dat_discrete=bs_dat.copy()  
    
    # Discretize the data and compute the set of specific bugs at the given cutoff
    cutoff=0.0
    
    # Discretize all but blood
    bs_dat_discrete[bs_dat_discrete>cutoff]=1
    bs_dat_discrete[bs_dat_discrete<=cutoff]=0
    bs_dat_discrete_noblood=bs_dat_discrete.loc[:,bs_dat_discrete.columns!='Blood']
    bs_dat_discrete_noblood=bs_dat_discrete_noblood.loc[:,bs_dat_discrete_noblood.columns!='CordBlood_Serum']
    
    # Examine threshold for discretization
    for i,site in enumerate(bs_dat_discrete_noblood.columns):   
        
        bs_dat.sort(site,inplace=True)
        load_bugs_in_site=bs_dat.loc[:,site]
        load_bugs_in_site=load_bugs_in_site[load_bugs_in_site>=0]
        load_bugs_in_site.sort()
        ind=np.arange(load_bugs_in_site.shape[0])
        plt.subplot(2,2,1)
        plt.bar(ind-0.45,np.log10(np.array(load_bugs_in_site)),color='k')
        plt.axhline(cutoff,linestyle='--',lw=1,c='red')
        plt.plot(bs_dat_discrete_noblood.loc[load_bugs_in_site.index,site],drawstyle='steps')
        # locs,labels=plt.xticks(ind,load_bugs_in_site.index)
        # plt.setp(labels, rotation=90, fontsize=3.5)
        # locs,labels=plt.yticks()
        # plt.setp(labels,fontsize=5)
        plt.ylabel('Mean fractional abundance (log10)',fontsize=6)
        plt.title('%s: Ranked abundance of bugs in: %s'%(type_dat,site),fontsize=6)
        plt.xlim([-1,load_bugs_in_site.shape[0]])
        plt.ylim([plt.ylim()[0],1.2])
        plt.show()
        
    return(bs_dat_discrete,bs_dat_discrete_noblood)

binary_16s,binary_16s_noblood=discretize(mean_data_16s,'16s')
```

```python
binary_hmp,binary_hmp_noblood=discretize(mean_data_hmp,'hmp')
```

```python
def site_specific_in_blood(makePlot,bs_data,site_spec):        
    only_blood=bs_data.loc[:,'Blood']
    in_blood=only_blood[only_blood>0]
    stor_specific={}
    spec_df=pd.DataFrame(index=list(site_spec.keys()),columns=['All Specific to bodysite','In blood'])
    for i,to_compute in enumerate(site_spec.keys()):
        # Present in body site
        in_bs=site_spec[to_compute]
        # Present elsewhere
        specific_bugs=list(set(in_bs)-set(in_blood.index))
        spec_df.loc[to_compute,'All Specific to bodysite']=len(in_bs)
        spec_df.loc[to_compute,'In blood']=len(in_bs)-len(specific_bugs)
        stor_specific[to_compute]=specific_bugs
        if makePlot:
            plt.subplot(2,2,1)
            venn2([set(in_bs), set(in_blood.index)],set_labels = ('In: %s'%to_compute, 'In: Blood'),set_colors=('red','blue'),alpha=0.5)
            c = venn2_circles(subsets=[set(in_bs), set(in_blood.index)], linestyle='solid')
            c[0].set_lw(1.5) 
            c[1].set_lw(1.5) 
            plt.show()
    return stor_specific,spec_df

in_tis_not_blood_hmp,spec_df_hmp=site_specific_in_blood(1,binary_hmp,site_specific_hmp)
```

```python
def plotload(bs_dat,dat_type,site_specific_bugs):
    stor_blood=pd.DataFrame(columns=['Load'])
    for site in site_specific_bugs.keys():
        stor_blood.loc[site]=bs_dat.loc[site_specific_bugs[site],'Blood'].sum()
    stor_blood.loc['Mixed']=bs_dat.loc[:,'Blood'].sum()-stor_blood.loc[:,'Load'].sum()
    stor_blood.sort('Load',inplace=True,ascending=False)
    ind=np.arange(stor_blood.shape[0])
    plt.bar(ind-0.45,np.array(stor_blood),color='k')
    locs,labels=plt.xticks(ind,stor_blood.index)
    plt.setp(labels, rotation=90, fontsize=3.5)
    locs,labels=plt.yticks()
    plt.setp(labels,fontsize=5)
    plt.xlabel('Specific source',fontsize=6)
    plt.ylabel('Aggregate load',fontsize=6)
    plt.title('%s: Aggregate load of bugs from different sources'%dat_type,fontsize=6)
    plt.xlim([-1,stor_blood.shape[0]])
```

```python
from itertools import chain

def makeHeatmap(pruned_DF):
    ''' Heatmap summary of results from regression.'''
    
    data=np.array(pruned_DF,dtype=float)
    sample_Labels=np.array(list(pruned_DF.index))
    bugLabels=np.array(list(pruned_DF.columns))
    
    p=plt.pcolormesh(data,cmap='gist_heat')
    
    locs,labels=plt.yticks(np.arange(sample_Labels.size)+0.5,sample_Labels)
    plt.setp(labels, fontsize=5)
    
    plt.ylim(0,len(sample_Labels))
    plt.tick_params(\
    axis='x',         
    which='both',      
    bottom='off',     
    top='off')         
    # labelbottom='off') 
    
    plt.tick_params(\
    axis='y',          
    which='both',      
    left='off',      
    right='off',        
    labelleft='on')  
    
    ind=np.arange(bugLabels.size)+0.5
    locs,labels=plt.xticks(ind,bugLabels)
    plt.setp(labels, rotation=90, fontsize=5)
    plt.xlim([0,len(bugLabels)]) 

def prepHeatmap(bs_dat,dat_type,site_specific_bugs):
    # Matrix that maps each bug to its specific site
    
    site_map=pd.DataFrame(index=list(bs_dat.index),columns=list(site_specific_bugs.keys()))
    for site in site_specific_bugs.keys():
        site_map.loc[site_specific_bugs[site],site]=1
    site_map.fillna(0,inplace=True)
    non_spec=set(bs_dat.index)-set(chain.from_iterable(site_specific_bugs.values()))
    site_map.loc[:,'Mixed']=0
    site_map.loc[non_spec,'Mixed']=1
    
    # Blood data 
    blood_data=bs_dat.loc[:,'Blood']
    blood_data_nonzero=blood_data[blood_data>0]
    blood_data_nonzero.sort()
    
    # Load per tissue
    blood_dat_map=pd.DataFrame(index=list(in_blood.index),columns=site_map.columns)
    for bug in blood_data_nonzero.index:
        blood_dat_map.loc[bug,:]=site_map.loc[bug,:]*blood_data_nonzero.loc[bug]
    blood_dat_map=blood_dat_map.loc[blood_data_nonzero.index,:]
        
    sum_cat=blood_dat_map.sum(axis=0)
    sum_cat.sort(ascending=False)
    blood_dat_map=blood_dat_map[sum_cat.index]
    
    # Number of bugs to show in the final heatmap
    rank=25
    blood_dat_map=blood_dat_map.loc[blood_dat_map.index[-rank:]]
    
    makeHeatmap(blood_dat_map)
    sum_cat=sum_cat.apply(lambda x: np.round(x,decimals=2))
    bugLabels=[s+' (%s)'%sum_cat.loc[s] for s in sum_cat.index]
    ind=np.arange(len(bugLabels))+0.5
    locs,labels=plt.xticks(ind,bugLabels)
    plt.xlim([0,len(bugLabels)])
    plt.setp(labels, rotation=90, fontsize=5)
    cb = plt.colorbar()
    ax = cb.ax
    ax.tick_params(labelsize=5) 
        
    plt.xlabel('Likely tissue source (fractional abundance)',fontsize=6)
    plt.title('%s: Mean fractional abundance of each %s in blood \n likely tissue sources highlighted'%(dat_type,taxLevel_for_analysis),fontsize=6)
    plt.tight_layout()

fig = plt.Figure()
    
plt.subplot(2,2,1)
plotload(mean_data_hmp,'HMP',site_specific_hmp)

plt.subplot(2,2,3)
prepHeatmap(mean_data_hmp,'HMP',site_specific_hmp)

plt.subplot(2,2,2)
plotload(mean_data_16s,'16s',site_specific_16s)

plt.subplot(2,2,4)
prepHeatmap(mean_data_16s,'16s',site_specific_16s)

#fig.tight_layout()
#fig.savefig(os.getcwd()+'/Desktop/Blood_Composition_%s.png'%taxLevel_for_analysis,format='png',bbox_inches='tight',dpi=150,pad_inches=0.5)
#fig.savefig(os.getcwd()+'/Desktop/Blood_Composition_%s.pdf'%taxLevel_for_analysis,format='pdf',bbox_inches='tight',dpi=150,pad_inches=0.5)
```

```python
# Run PCA
import brewer2mpl
from sklearn.decomposition import PCA

def PCApreprocess(df):
    ''' Zero out the mean, and made stv one. '''
    # Normalize data by samples with zero mean 
    df_mean_norm=df.subtract(df.mean(axis=1),axis=0)
    # Normalize data by samples with unit variance 
    df_mean_var_norm=df_mean_norm.divide(df_mean_norm.std(axis=1),axis=0)
    return df_mean_var_norm

def getLoadings(df,pca,n_comp):
    loadings = pca.components_
    storeLoad=pd.DataFrame(index=list(df.columns),columns=[range(n_comp)])    
    for i in range(n_comp):
        storeLoad[i]=loadings[i]
    
    for i,load in enumerate(storeLoad.columns):
        # ind=np.arange(storeLoad.shape[0])
        plt.subplot(2,1,i+1)
        storeLoad.sort(load,inplace=True)
        # plt.bar(ind,np.array(storeLoad[load]),color='black')
        l=pd.DataFrame(storeLoad[load])
        tmp=pd.concat([l.head(10),l.tail(10)],axis=0)
        ind=np.arange(tmp.shape[0])
        plt.bar(ind,np.array(tmp),color='black')
        locs,labels=plt.xticks(ind+0.5,tmp.index)
        plt.setp(labels, rotation=90, fontsize=4)
        locs,labels=plt.yticks()
        plt.setp(labels,fontsize=5)
        plt.ylabel('Feature loading',fontsize=5)
        plt.title('PC %s: 10 lowest and 10 highest microbe loadings'%(i+1),fontsize=5)
        plt.xlim([0,tmp.shape[0]])
        plt.show()

# Preprocess
df_out=PCApreprocess(dat_tp_norm)
df_out.fillna(0,inplace=True)

# Index to show PCA data
data_index_toslice=df_out.index.get_level_values('BodySite')

# PCA
n_comp = 2
pca = PCA(n_components=n_comp)
siteData_r = pca.fit(df_out).transform(df_out)

# Plot
colorMapMax = 8
bmap = brewer2mpl.get_map('Paired', 'qualitative', colorMapMax).mpl_colors
uniq_ix=list(set(data_index_toslice))
plt.figure()
for ix,c in zip(uniq_ix,['r','g','b','c','k','y','m']):
    plt.scatter(siteData_r[data_index_toslice==ix, 0], siteData_r[data_index_toslice==ix, 1], facecolor=c, edgecolor='black', label=ix, alpha=1)
plt.legend(loc='upper right',scatterpoints=1,prop={'size':4})
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
plt.xlabel('PC1',fontsize=5)
plt.ylabel('PC2',fontsize=5)
title='All body sites'
plt.title('PCA of %s data'%title,fontsize=5)
plt.show()

# Get the loadings
getLoadings(df_out,pca,n_comp)
```

```python
import scipy
from scipy import stats 
from scipy.stats import gaussian_kde

def getBugData(bug):
    ''' Cohort data for a specific bug with (1) binary blood data and (2) matched body site data. '''
    # Isolate cohort data for the bug of interest
    bug_dat=dat_tp_norm[bug]
    bug_dat=bug_dat.unstack()
    # Pick most commonly assayed sites
    to_sample=['Blood','Saliva','Stool','Tooth_Gum','Vaginal_Swab']
    bug_dat=bug_dat[to_sample]
    # Drop any sample with missing data for one of the body sites
    bug_dat=bug_dat.dropna(axis=0)
    # All body sites listed in the resulting data frame have a complete set of samples
    bug_dat_binary=bug_dat.transpose()
    bug_dat_binary.loc['Blood',bug_dat_binary.loc['Blood',:]>0]=1
    return bug_dat_binary
```

```python
# Consider all bugs in blood
bugsToEval=list(bugs_found_in_blood.index)
attributes=['Saliva','Stool','Tooth_Gum','Vaginal_Swab']
storeP=pd.DataFrame(index=bugsToEval,columns=attributes)

# Compute P-values
for bug in bugsToEval:
    # Bug data
    bug_dat_binary=getBugData(bug)
    # Iterate through attributes
    for tissueToEval in attributes:
        blood_tissue_comp=pd.DataFrame()
        blood_tissue_comp['Tissue_Load']=bug_dat_binary.loc[tissueToEval,:]
        blood_tissue_comp['In_Blood']=bug_dat_binary.loc['Blood',:]
        blood_tissue_comp.set_index('In_Blood',inplace=True)
        try:
            not_in_blood=blood_tissue_comp.loc[0,'Tissue_Load']
            in_blood=blood_tissue_comp.loc[1,'Tissue_Load']
            wilcox=scipy.stats.ranksums(not_in_blood,in_blood) 
            storeP.loc[bug,tissueToEval]=wilcox[1]
        except:
            # print "Cannot compute for %s."%bug
            None
storeP.fillna(1,inplace=True)

# Plot
plt.subplot(1,2,1)
sum_tis=storeP.sum(axis=0)
sum_tis.sort()
sum_bug=storeP.sum(axis=1)
sum_bug.sort()
storeP=storeP.loc[sum_bug.index,sum_tis.index]

# Omit cases in which no P-value is computed (due to no bug found in tissues).
p_scores=storeP.sum(axis=1)
valid=p_scores[p_scores!=4]
storeP_toShow=storeP.loc[valid.index]
makeHeatmap(storeP_toShow)
plt.colorbar()
plt.title('Wilcox test P-values',fontsize=6)
```

```python
from scipy import stats
import rpy2
import rpy2.robjects as robjects
%load_ext rmagic 
```

```python
import random

def runQP(test_sample,blindSource):
    '''Run QP in R'''
      
    # Ensure cord blood "tissue" not included
    try:
        test_sample=test_sample[test_sample.index!='CordBlood_Serum']
    except KeyError:
        None
        
    # Add blind source
    if blindSource:
        test_sample.loc['Blind_Source']=0
        
        # Blind source
        test_sample.loc['Blind_Source','Cupriavidus']=test_sample.loc['Blood','Cupriavidus']
        test_sample.loc['Blind_Source','Pseudomonas']=test_sample.loc['Blood','Pseudomonas']
        
        # Add back obvious skin bugs
        test_sample.loc['Blind_Source','Propionibacterium']=test_sample.loc['Blood','Propionibacterium']
        test_sample.loc['Blind_Source','Staphylococcus']=test_sample.loc['Blood','Staphylococcus']
                
    # Model parameters
    B=test_sample.loc['Blood']
    B_sites=test_sample.drop('Blood')
    B_sites=B_sites.transpose()
    tissues=len(B_sites.columns)
    Dmat_in=np.dot(B_sites.transpose(),B_sites)
    dvec_in=np.dot(B.transpose(),B_sites)
    
    # Push parameters into R and run the QP model 
    %Rpush B B_sites tissues dvec_in Dmat_in
    
    # Pull in this package
    %R require(quadprog)
    
    # Body site matrix and blood target
    %R X <- B_sites
    %R Y <- B
    
    # Remove constraint that all tissues must sum to one! 
    # QP constraints and constraint target values 
    #%R C <- cbind(rep(1,tissues),diag(tissues))
    %R C <- cbind(diag(tissues))
    #%R b <- c(1, rep(0, tissues))
    %R b <- c(rep(0, tissues))
    #%R sol <- solve.QP(Dmat = Dmat_in, dvec = dvec_in, Amat = C, bvec = b, meq = 1)
    %R sol <- solve.QP(Dmat = Dmat_in, dvec = dvec_in, Amat = C, bvec = b, meq = 0)
    %Rpull sol
    
    # Model
    model_coef=sol[0]
    
    return (B_sites,model_coef,B,sol)
```

```python
# Define sample to test
sample='PR10047_27'
test_sample=dat_tp_norm.loc[sample]
# Blood data
no_blood=test_sample[test_sample.index!='Blood']
# Random weights
rand_nums=np.array([random.uniform(0,1) for x in range(4)])
tissue_weight=rand_nums/rand_nums.sum()
# Generate blood data
for bug in set(no_blood.columns):
    test_sample.loc['Blood',bug]=np.sum(no_blood[bug]*tissue_weight)
# Run QP and show analysis plot
B_sites,model_coef,B,all_model_results=runQP(test_sample,0)
# Scatter
plt.subplot(2,2,1)
plt.scatter(tissue_weight,model_coef,color='k',s=5)

model_guess=(B_sites*model_coef).sum(axis=1)
slope, intercept, r_value, p_value, std_err = stats.linregress(B,model_guess)
r2=r_value**2

#r2,model_guess,df=analyzeSample(B_sites,model_coef,B)
regression = np.polyfit(np.array(tissue_weight),np.array(model_coef), 1)
xvals=np.arange(plt.xlim()[0],plt.xlim()[1],0.01)
plt.plot(xvals,regression[0]*xvals+regression[1],lw=0.5,linestyle='--',alpha=0.5,color='k')
plt.xlabel('Tissue weights used',fontsize=5)
plt.ylabel('Tissue weights recovered',fontsize=5)
plt.title('Test QP model with mock data with $r^2=$%s'%r2,fontsize=5)
```

Iterate through all samples, noting that we only evaluate samples with > body sites sampled.

```python
# Samples
samples=list(set(dat_tp_norm.index.get_level_values('Sample')))
storCoef=[]
storResiduals=[]
r2vals=pd.DataFrame(columns=['r2'])
t=4

# Iterate through samples
for sample in samples:
    # Only evaluate samples with > t body sties     
    if dat_tp_norm.loc[sample].shape[0] > t:
        # Run QP
        test_sample=dat_tp_norm.loc[sample]
        B_sites,model_coef,B,all_model_results=runQP(test_sample,0)
        
        # Store model data 
        df=pd.DataFrame()
        df['Per']=np.array(model_coef)
        df['Body sites']=B_sites.columns
        df.set_index('Body sites',inplace=True)
        df.sort('Per',inplace=True)
        model_guess=(B_sites*model_coef).sum(axis=1)
        slope, intercept, r_value, p_value, std_err = stats.linregress(B,model_guess)
        # Store
        df.loc['r2']=r_value**2
        df.loc['Sample']=sample
        storCoef=storCoef+[df]
        
        # Store data
        #delta=pd.DataFrame(np.abs(model_guess.subtract(B,axis=0)),columns=[sample])
        delta=pd.DataFrame(model_guess.subtract(B,axis=0),columns=[sample])
        storResiduals=storResiduals+[delta]
        r2vals.loc[sample,'r2']=r_value**2

# Concatenate
all_res=pd.concat(storResiduals,axis=1)
stor_cocat=pd.concat(storCoef,axis=1)
```
