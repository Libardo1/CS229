## Supervised learning-

### General objective - 

The objective of supervised learning is to $learn$ a function, $h :x \mapsto y $, so that $h(x)$ is a good predictor of $y$. 

$x$ is typically called a feature.

When $y$ is continuous, this is a $\textbf{regression}$ problem. When $y$ is categorical (discrete), this is a $\textbf{classification}$ problem. 

For a simple regression problem, consider records measurements of 13 attributes of housing markets around Boston, as well as the median price. 

The question is can you predict the price of a new market given its attributes?

```python
from sklearn.datasets import load_boston
boston = load_boston()
plt.hist(boston.target)
plt.xlabel('price ($1000s)')
plt.ylabel('count')
```

$x$ is often used to denote the input variables (features) whereas $y$ is the output (target). 

A pair $(x,y)$ is one training example and there are $m$ training examples.

$\textbf{Feature selection}$ is a simple, manual way to evaluate the relationship between the measured features and target.

But, we ultimatly want a funtion, $h$, do capture this relationship and predict the target formally.

### Linear regression-

$\textbf{Linear regression}$ approximates $y$ as a linear funtion of $x$: $y = h(x) = \sum\limits_{i=0}^n \theta_i x_i = \theta^T x $ where $\theta^T$ and $x$ are vectors.

In the following example, the vector of coefficients ($\theta^T$) are assigned to the vetor of features ($x$) when the "training data" is fit.

```python
# Use linear regression package.
from sklearn.linear_model import LinearRegression
clf = LinearRegression()

# When we fit the observed features to target, we determine the coefficients used in our linear funtion or model. 
clf.fit(boston.data,boston.target)
# Pass the features for each point to the funtion, resulting in a prediction.
predicted=clf.predict(boston.data)
# Examine how well the prediction matches the data.
plt.scatter(boston.target,predicted)
plt.plot([0, 50], [0, 50], '--k')
plt.axis('tight')
plt.xlabel('True price ($1000s)')
plt.ylabel('Predicted price ($1000s)')
```

#### A general cost model finding our coefficients - 

One way of doing this is defining a cost funtion that chooses coefficients $\theta$ such that $h(x)$ is close to $y$ for the training examples in the set of $m$ training examples.

$J(\theta) = \frac{1}{2} \sum\limits_{i=1}^m (h_{\theta} (x^i) - y^i)^2$

To be clear, $h_{\theta}$ is the linear funtion of the features, $x$, parameterized by the coefficients $\theta$.

Thus, come up with a $\theta$ that will minimize the squared distance (or squared error) between the prediction and the target for the training set.

This cost model is used in the $\textbf{ordinary least squares}$ regression model.

This is a special case in a broader family of algorithms (linear models).

#### An algorithm for finding the coefficients that minimize the cost model, J() - 

$J(\theta)$ is convex quadratic function, which means that it has a single local minimum.

We need a search algorithm to find $\theta$ that will minimize squared error and find the $\theta$ corresponding to this local minimum.

$\textbf{Gradient descent}$ is one algorithm for doing this.

It starts with a guess for $\theta$ (a vector) and repeatdly changes this vector in order to reduce $J(\theta)$ and, eventually, minimize the cost function.

It does this by taking steps in the direction that allows you to move downhill and reduce $J(\theta)$ as rapidly as possible.

As this proceeds, values of $\theta$ and - correspondingly - $J(\theta)$ will change and eventually converge to a local optimum.

With $\alpha$ as the learning rate (step size parameters) and := indicating assignment of values (versus = being an assertion of equality):

$ \theta_j := \theta_j - \alpha \frac{\partial}{\partial \theta_j} J(\theta) $ where the partial derivative of $J(\theta)$ with respect to $\theta_j$ indicates the direction of steepest descent of the function.

As you can see, $\alpha$ will set the step size, or the length of the step taken in the direction of greatest descent.

Also, the gradient - $\frac{\partial}{\partial \theta_j} J(\theta)$ - will get smaller as it approaches, and reaches zero at, the local minima. Thus, the value of $\alpha$ does not need to be changed.

To implement $\textbf{Gradient descent}$ for linear regression, we solve $\frac{\partial}{\partial \theta_j} J(\theta)$.

After some math (see lecture notes), we obtain the $\textbf{least mean squares (LMS) updater rule}$:

(1) For any training example - $ \theta_j := \theta_j + \alpha (y^i - (h_{\theta} (x^i)) x_j^i $ for a single training example.

(2) Batch (for all training examples) - $ \theta_j := \theta_j + \alpha  \sum\limits_{i=1}^m (y^i - h_{\theta} (x^i)) x_j^i $ for all training examples.

Intuitivly, $(1)$ says that the features for a training example $x_j^i$ are mutiplied by the difference between the prediction given $\theta$ and the taget $(y^i - (h_{\theta} (x^i))$. 

As $\theta$ gets better and the prediction is closer to the target, the difference approaches zero $\theta$ is not updated: the magnitude of the update is proportional to the error.

```python
# This is an implementation of stochastic gradient descent.

def lms(data,eta,iterations):

    # Get the number of attributes/features
    target = data[1]
    data = data[0]
    dim = data.shape[1] 
    num_points = data.shape[0]
    
    # Starting guess for theta.
    w = np.ones(dim) # that is the weight vector as column vector   
    print "Starting theta: %s"%w

    # Number of training examples to sample.
    for i in range(iterations):
        
        # Pick a sample from the training data.
        next = np.random.randint(num_points)  
        
         # Make a prediction based upon the current theta (w).
        predict = np.dot(data[next,],w) 
        
        # Determine the error.
        error = target[next]-predict       
        
        # Increment theta: training example * error * step size + pre-existing guess.
        w = w + eta*error*data[next,]  
        
    print "Theta after %s iterations: %s"%(iterations,w)
    return w
```

```python
# Set of training examples.
m=10
# Number of features and the actual coefficient used to produce target data.
f=1
theta=0.2
toy=np.arange(m).reshape((m,f))
# Make target data
target=toy*theta
```

```python
# Set iterations used for stochastic gradient descent.
it=15
eta=.05
plt.subplot(1,2,1)
data_zipped=(toy,target)
w=lms(data_zipped,iterations=it,eta=eta)
plt.plot(toy,target,'g.',markersize=20,alpha=0.5)
plt.plot(toy,np.dot(toy,w),'r.',markersize=20,alpha=0.5)
plt.show()
data_zipped = (toy,target)
```

#### An explicit way to find coefficients that minimize the cost model, J() - 

Remember, our objective is to fine $\theta$ that minimizes $J()$.

Rather than the LMS updater rule, we'll find a closed-form solution. 
    
Recall gradient of a matrix:
    
\begin{equation} (\nabla_A f(A))_{ij} = \frac{\partial(A)}{\partial(A_{ij})} \end{equation}

Given a matrix:

\begin{equation} A= \begin{bmatrix} A_{11} & A_{12} \\ A_{21} & A_{22} \end{bmatrix} \end{equation}

And a funtion applied to the matrix that returns a scalar:

\begin{equation} f(A) = \frac{3}{2}A_{11} + 5A_{12}^2+A_{21}A_{22} \end{equation}

Then:

\begin{equation} \nabla_A f(A) = \begin{bmatrix} \frac{3}{2} & 10A_{12} \\ A_{22} & A_{21} \end{bmatrix} \end{equation}

With this in mind, let's re-write J():

First, we define the $\textbf{design matrix}$, $X$, that contains traning example input values as rows as well as a vector, $y_{tar}$, with the target values.

With some math, the cost equation can be re-cast in terms of $X$ and $y$: 

$J(\theta) = \frac{1}{2} \sum\limits_{i=1}^m (h_{\theta} (x^i) - y^i)^2 = \frac{1}{2} (X\theta - y_{tar})^T (X\theta - y_{tar}) $

$J(\theta)$ is a funtion of the training set matrix, $X$, as well as the targets $y_{tar}$.

With respect to the example above, the cost funtion is applied to the varaible $\theta$.

We then compute the derivative:

\begin{equation} \nabla_{\theta} J(\theta) \end{equation}

After some math, you get the following:

\begin{equation} \nabla_{\theta} J(\theta) = X_T X \theta - X^T y_{tar} \end{equation}

Set this to zero in order get the value of $\theta$ at which there's no change in the cost funtion, indicating convergence.

Thus, the value of $\theta$ that minimizes $J(\theta)$:

\begin{equation} \theta = (X_T X)^{-1} X^T y_{tar} \end{equation}
