```python
# Pull data for cohort, which is used to compute percentiles.
cohort_dat=sql.read_sql("select Grammy_Table.Sample_ID,Tax_Table.%s,SUM(Grammy_Table.Normalized_Abundance) "
                 "from Tax_Table "
                 "INNER JOIN Grammy_Table "
                 "ON Tax_Table.tax_id=Grammy_Table.TaxID "
                 "WHERE Grammy_Table.Study in (%s)"
                 "GROUP BY Grammy_Table.Sample_ID,Tax_Table.%s "%(level,q,level),engine)
cohort_dat=cohort_dat[~pd.isnull(cohort_dat[level])]
cohort_dat[level]=cohort_dat[level].astype(int)
cohort_dat.set_index(['sample_id','%s'%level],inplace=True)
c=cohort_dat.unstack()['sum'].fillna(0)
# Pull data for cohort, which is used to compute percentiles.
s=sql.read_sql("select * from sample_table",engine)
s=s[['daysposttransplant','sample_id','patient_id']]
s.set_index('sample_id',inplace=True)
m=pd.merge(c,s,right_index=True,left_index=True)
m.reset_index(drop=False,inplace=True)
m.head(3)
```

```python
# Clinical data
clinical_data=sql.read_sql("select * from Clinical_Table",engine) 
clinical_data.fillna('',inplace=True)
clinical_data.set_index('index',inplace=True)
clinical_data.head(3)
```

```python
# Tax encoding
clinical_tax_encode=sql.read_sql("select * from Clinical_Tax_Encoding_Table",engine) 
clinical_tax_encode.set_index('index',inplace=True)
clinical_tax_encode.head(3)
```

```python
# Inputs
infectionKey='CMV'
infectionTaxID=clinical_tax_encode.loc[infectionKey,level]
proximityThreshold=10
```

```python
def replaceZeros(DF,infectionTaxID):
    ''' Replace zero genome equivalent values with detection limit. '''
    # Replace zero genome equivalent values with detection limit 
    d=sql.read_sql("select * from Grammy_Table WHERE Grammy_Table.Study in (%s)"%(q),engine)
    defInfGra=d[d.compressed.str.contains(','+str(int(infectionTaxID))+',')] # All data for given infection
    meanGenome=np.mean(defInfGra['genome_size']) # Mean genome size for all taxIDs under the specified infection taxID
    s=sql.read_sql("select * from stat_table",engine)
    detLimit=float(1)/float(meanGenome)/np.mean(s['chr21_coverage']) # 1 read / genome size / sample sample coverage for cohort. 
    DF.loc[DF[infectionTaxID]==0,infectionTaxID]=detLimit # Replace zero values
    return DF

t_tests=clinical_data.loc[infectionKey,:]
try:
    num=t_tests.shape[1]
# Correct for only one test found.
except IndexError: 
    t_tests=pd.DataFrame(t_tests) 
    t_tests=t_tests.transpose()        
    
t_measure=m[['patient_id','sample_id','daysposttransplant',infectionTaxID]]
data=np.array(t_measure[infectionTaxID])
t_measure['percentile']=[stats.percentileofscore(data, i, kind='strict') for i in data]
t_merge=pd.merge(t_tests,t_measure,left_on='patient_id',right_on='patient_id',how='inner')
t_merge['daysposttx_test']=t_merge['daysposttx_test'].astype(float)
t_merge['daysposttransplant']=t_merge['daysposttransplant'].astype(float)
t_merge['delta']=t_merge['daysposttx_test'].sub(t_merge['daysposttransplant'])
t_merge['delta']=np.abs(t_merge['delta'])
t_merge=t_merge[t_merge['delta'] < proximityThreshold]
t_merge.set_index('level_0',inplace=True)
t_merge=replaceZeros(t_merge,infectionTaxID)
t_merge.head(3)
```

```python
# --- Re-cast some positive values ---
blacklist1=[' CMV (<150)','CMV (<150)','CMV(>400)','CMV (DETECTED)','CMV (<135)',' CMV (<135)','CMV (<174)'] # CMV data, lung
blacklist2=['CMV (<135; <2.13)','CMV (<135; log 10-<2.13)','CMV (<135 IU/mL <2.13 Log 10)'] # CMV data, lung
blacklist3=['ADENOVIRUS (results <120; result log 10-<2.08)','ADENOVIRUS ()','ADENOVIRUS (<120; result log 10-<2.08)','ADENOVIRUS (<120; <2.08)','ADENOVIRUS (<120; result log 10<2.08)']
blacklist4=['HHV6 ()','HHV6 (DETECTED)','HHV6 (<1000 copies/ml)','HHV6 (<1000; log 10-<3.00)','HHV6 (<1000; result log 10- <3.00)','HHV6 (<1000; result log 10 <3.00)','HHV6 (<1000; <3.00)','HHV6 (result <1000; result log 10 <3.00)']
blacklist=blacklist1+blacklist2+blacklist3+blacklist4

for falsePosResult in blacklist:
    t_merge.loc[t_merge['detail']==falsePosResult,'test_result']=0  

# Enforce tests done in body fluid
bodySite='Serum' # CMV
t_merge=t_merge[t_merge['specimen']==bodySite]
```

```python
from scipy import stats 
from scipy.stats import gaussian_kde
from sklearn.metrics import roc_curve, auc

def makeKDE(dfToRun,toPlot,result,color,label,offset):
    dat=dfToRun[dfToRun['test_result']==result]
    if toPlot!='percentile':
        sample=np.log10(np.array(dat[toPlot]))
        x=np.linspace(np.min(sample)*1.1,np.max(sample),50)
    else:
        sample=np.array(dat[toPlot])
        x=np.linspace(-10,110,50)
    density=stats.kde.gaussian_kde(sample)   
    plt.plot(x,density(x),c=color,label=label)
    plt.plot(sample,[offset*plt.ylim()[1]]*len(sample),'|',color=color)
    plt.xlabel('%s'%toPlot,fontsize=5)
    plt.ylabel('Density',fontsize=5)
    plt.title('Kernel desnity estimate.',fontsize=5)

def dispKDEs(dfToRun,toPlot):
    result=1
    color='blue'
    makeKDE(dfToRun,toPlot,result,color,'With report',0.5)
    result=0
    color='red'
    makeKDE(dfToRun,toPlot,result,color,'Without report',0.5)
    plt.legend(prop={'size':5},loc='upper right')
    plt.xticks(fontsize=5)
    plt.yticks(fontsize=5)
    
def makeBoxPlot(DF,infectionTaxID,key,spec,DB,colToEval):
    withoutClinicalReport=DF[DF['test_result']==0]
    withClinicalReport=DF[DF['test_result']==1]
    woRep=np.array(withoutClinicalReport[infectionTaxID])
    wRep=np.array(withClinicalReport[infectionTaxID])
    if colToEval != 'Percentile':
        data=[np.log10(woRep),np.log10(wRep)] 
    else:
        data=[woRep,wRep] 
    plt.boxplot(data)
    wilcox=scipy.stats.ranksums(withoutClinicalReport[infectionTaxID],withClinicalReport[infectionTaxID]) 
    labels = ('Without \n report', 'With \n report')
    plt.xticks(range(1,3),labels, rotation=0,fontsize=5)
    plt.ylabel('%s'%colToEval,fontsize=5)
    locs,labels=plt.xticks()
    plt.setp(labels, fontsize=5)
    locs,labels=plt.yticks()
    plt.setp(labels,fontsize=5)
    try:
        plt.title('Measure %s for %s via %s database \n NumData_neg=%s, NumData_pos=%s \n -log10 P-value=%s \n P-val=%s \n Data type=%s'%(getName(infectionTaxID),spec,key,DB,woRep.shape[0],wRep.shape[0],-1*np.log10(wilcox[1]),wilcox[1],colToEval),fontsize=5)
    except:
        plt.title('Measure %s for %s via %s database \n NumData_neg=%s, NumData_pos=%s \n -log10 P-value=%s \n P-val=%s \n Data type=%s'%(spec,key,DB,woRep.shape[0],wRep.shape[0],-1*np.log10(wilcox[1]),wilcox[1],colToEval),fontsize=5)
    if colToEval == 'Percentile':
        plt.ylim([plt.ylim()[0]+-10,plt.ylim()[1]*1.1])
    return (wilcox[1],wRep.shape[0],woRep.shape[0])

def makeROC(DF,infectionTaxID,c):
    test_intNorm=DF[infectionTaxID]/np.max(DF[infectionTaxID])
    goldStd_intBinary=np.array(DF['test_result'],dtype=int)
    fpr, tpr, thresholds = roc_curve(goldStd_intBinary,test_intNorm)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr,tpr,lw=1.5,color=c)
    return (goldStd_intBinary,test_intNorm,fpr,tpr,thresholds,roc_auc)
```
