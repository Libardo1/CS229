### Introduction

#### Law of large numbers - 

The average of a sequence of random variables from the same distribution converges to the expected value of that distribution.

$$ \frac{1}{N} \sum\limits_{i=1}^N Z_{i} \rightarrow E[Z], N \rightarrow \infty $$

Let's sample 100000 Poisson random variables with parameter $\lambda$=4.5. The expected value of a Poisson random variable is equal to it's parameter (explained below).

```python
# Create a random sample 
sample_size = 100000
lambda_ = 4.5
N_samples = range(1, sample_size, 100) # 1000 samples
colors = matplotlib.rcParams['axes.color_cycle']
plt.subplot(2,1,1)
for k,c in zip(range(3),colors):
    samples = np.random.poisson(lam=lambda_, size=sample_size) # Draw samples from a Poisson distribution.
    partial_average = [samples[:i].mean() for i in N_samples] # Draw increacingly large N 
    plt.plot(N_samples, partial_average, lw=2, color=c, label="Average of  $n$ samples; Trial: %d" % k)
plt.plot(N_samples, lambda_ * np.ones_like(partial_average),ls="--", label="Expected value", c="k")
plt.ylim(4.35, 4.65)
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
plt.title("Convergence of the average of random variables to its expected value",fontsize=7.5)
plt.ylabel("Sample mean for $n$ samples",fontsize=5)
plt.xlabel("# of samples, $n$",fontsize=5)
plt.legend(loc='upper right', scatterpoints = 1,prop={'size':5})
plt.grid()
```

For small N (few samples drawn), the mean value deviates from the expected value. For large N, the sample mean approaches the expected value, $\lambda$.

### Probability Distributions 

(1) Mean is the expected value and is influenced by outliers. (2) The median is that value that comes half-way when the data are ranked in order. In contrast to the mean, it is not affected by outlying data points. (3) The mode value is the most frequently occurring value in a distribution. (4) Range is the difference between the highest and the lowest data value. (5) Cumulative distribution function (CDF) tells you for each value which percentage of the data has a lower value: for example, the median is the 50% percentile and the first quartile (Q1) is the 25% percentile (25% of the data has a lower value). (6) The variance, with n−1 giving a better estimate of the population variance:

$$\sum\limits_{i=1}^n ( x_{i}-\overline{x} )^2 / (n-1) $$

(7) Standard deviation, $\sigma$, is the square root of the variance. (7) Standard error is $\sigma / \sqrt{n}$, with n being the number of samples.

#### Discrete distributions: Binomial -

The distribution of a discrete random variable can be characterized through its probability mass function (pmf). The probability that a given number will be the realization of a discrete random variable is equal to the value taken by its pmf in correspondence of that number. 

```python
# Parameter sweep of successes
toTest=10
p=0.5
N_vals=[2,5,10]
n=np.arange(0,toTest)
colors = matplotlib.rcParams['axes.color_cycle']

# Iterate through scenarios
plt.subplot(2,2,1)
for N,c in zip(N_vals,colors):
    y=stats.binom.pmf(n,N,p)
    plt.plot(n,y,'-',linewidth=2,color=c,alpha=1,label='N =%i'%N)
    plt.fill_between(n,y,color=c,alpha = .25)
    plt.xticks(fontsize=5)
    plt.yticks(fontsize=5)
    plt.title('Probability mass funtion',fontsize=7.5)
    plt.xlabel('n',fontsize=5)
    plt.ylabel('P(n)',fontsize=5)
    plt.legend(loc='upper right', scatterpoints = 1,prop={'size':5})
    plt.grid()

 # Iterate through scenarios
plt.subplot(2,2,2)
for N,c in zip(N_vals,colors):
    y=stats.binom.cdf(n,N,p)
    plt.plot(n,y,'-',linewidth=2,color=c,alpha=1,label='N =%i'%N)
    plt.fill_between(n,y,color=c,alpha = .25)
    plt.xticks(fontsize=5)
    plt.yticks(fontsize=5)
    plt.title('Cumulative distribution funtion',fontsize=7.5)
    plt.xlabel('n',fontsize=5)
    plt.ylabel('P(< or = n reads/base)',fontsize=5)
    plt.legend(loc='upper right', scatterpoints = 1,prop={'size':5})
    plt.xlim([0,plt.xlim()[1]])
    plt.ylim([0,plt.ylim()[1]])
    plt.grid()

plt.tight_layout()	
```

Considering a random variable, n, compute the probability of observing n successes for N trials given a probability P for each without regard for order. If N=5, p=0.5, n=2, then the first success can occour in any of the 5 trials. The second success can occour in any of the remaining 4 trials, giving 5 $\times$ 4 permutations. But, we don't care about the order so 5 $\times$ 4 / n! = 10 combinations of 2 successes. The number of combinations that satisfy the condition, n, is generally given by the binomial coefficient in the below probability mass function. This states: (1) there are N choose n combinations with (2) the probability of each combination $p^n (1-p)^{N-n}$ (n successes, N-n failures)

$$ P(n) = {N \choose n} p^n (1-p)^{N-n} $$

To illustrate, we can consider the case of reads and chromosomes. Given 10 reads that map to a chrom, what is the probabillity that 3 come from one copy (equal odds of mapping). Again, we can compute the numnber of combinations of three reads on a specific copy (above) and, using a Binomial probability distribution, can compute the probability of n reads out to 10 coming from one chrom copy. Note that the expected value of the binomial (mean) is Np, with n the number of trials and p the probability of success for each.

$$ {N \choose n} = \frac{N!}{n!(N-n)!} $$

Use $N=5$ and $n=2$ to explain Binomial coefficient. $n=2$ refers to two events (e.g., two heads). Given $N=5$ trials and that two have a head, the first head can occour at any one of the five but the second head is then restricted to any one of the remaining four. This results in 20 permutations of two heads in 5 trials. Of course, we don't care about the order: if we draw a 5x5 grid it is clear that we can omit half of it, resulting in 10 unique combinations of two heads, 

#### Discrete distributions: Poisson -

Assuming an expected value of a Binomial random variable, $\lambda$, let's assume $\lambda$ is a measured value such as the number of cars observed per hour. If we let N=60 (one trial per minute over the course of the hour), then we can compute the probability of success per trail, p=$\frac{\lambda}{60}$. If we want to compute the number of cars, n, obsered per minute then we again use the binomial where the probability of observing a car per minute is $\frac{\lambda}{60}$: 

$$ P(n=k) = {60 \choose k} \frac{\lambda}{60}^k (1-\frac{\lambda}{60})^{60-k} $$


However, we can have > 1 car per minute. We can get more granular. So, let's assume the number of trials is large (e.g., sample at >> second frequency) and the probability of success at each is correspondingly smaller. As N approaches infinity, Binomal becomes the Poisson distribution:

$$ \lambda = Np $$ 
$$ P(X=k) = \lim _{N \to \infty} {N \choose k} \frac{\lambda}{N}^k (1-\frac{\lambda}{N})^{N-k} = \frac{\lambda^k}{k!} e^{-\lambda} $$

```python
# The application to genomics is useful to consider
toTest=50 # Number of base position
n=np.arange(0,toTest) # Bases to examine
l=100 # Read length
N=3*10**9 # Size of human genome
reads=[10**8,10**8.5,10**9] # Sequencing reads
coverage=[float(read_number)*l/N  for read_number in reads] # Expected value (reads per base)
colors = matplotlib.rcParams['axes.color_cycle']

# Iterate through scenarios
plt.subplot(2,2,1)
for u,c in zip(coverage,colors):
    y=stats.poisson.pmf(n,u)
    plt.plot(n,y,'-',linewidth=2,color=c,alpha=1,label='$\lambda$ =%.3f'%u)
    plt.fill_between(n,y,color=c,alpha = .25)
    plt.xticks(fontsize=5)
    plt.yticks(fontsize=5)
    plt.title('Probability mass funtion',fontsize=7.5)
    plt.xlabel('n (base)',fontsize=5)
    plt.ylabel('P(n reads/base)',fontsize=5)
    plt.legend(loc='upper right', scatterpoints = 1,prop={'size':5})
    plt.grid()

plt.subplot(2,2,2)
for u,c in zip(coverage,colors):
    y=stats.poisson.cdf(n,u)
    plt.plot(n,y,'-',linewidth=2,color=c,alpha=1,label='$\lambda$ =%.3f'%u)
    plt.fill_between(n,y,color=c,alpha = .25)
    plt.xticks(fontsize=5)
    plt.yticks(fontsize=5)
    plt.title('Cumulative distribution funtion',fontsize=7.5)
    plt.xlabel('n (base)',fontsize=5)
    plt.ylabel('P(< or = n reads/base)',fontsize=5)
    plt.legend(loc='upper right', scatterpoints = 1,prop={'size':5})
    plt.grid()

plt.tight_layout()	
```

The total interval is an expariment and the reads arriving (spanning) a base is modeled as a Poisson random variable: N is large, as there are many reads of length l, but the probability of each read arriving at the given base is low. The expected number of reads per base is $Np=\lambda=N(l/L)$. This is the coverage and represents the expected number of reads per position. As coverage increaces, the PMF shows that the probability if getter a higher number of reads per base increaces. Of course, the probability of a specific position having *exactly* 30 reads is low < 0.1. But the CDF shows that the probability of getting 30 or more reads spanning the base when $\lambda=30$ is reasonable at 0.524 (1-y[29]).

#### Continuous distributions: Exponential -

A specific condition for Poisson is that (1) the time (distance) between successes (reads) is exponentially distributed with rate parameter $\lambda$ and (2) each trial (read) is independent. The probability of observing an event over an interval (of time, or in the case of sequencing, distance) is proportional to the length of that interval with $\lambda$ the constant of proportionality. The probability density funtion is, and it decays in proportion to the value of $\lambda$ and x, the length of the interval.

$$ P(x) = \lambda e^{-\lambda x} $$

```python
# Values
x=np.arange(0,4,0.1)
colors = matplotlib.rcParams['axes.color_cycle']
lambda_ = [0.5,1,2]

plt.subplot(2,2,1)
u_vals=[2,1]
for u,c in zip(coverage,colors):
    y=u*np.exp(-u*x)
    plt.plot(x,y,'-',linewidth=2,alpha=1,color=c,label='$\lambda$ =%.3f'%u)
    plt.fill_between(x,y,color=c,alpha = .25)
plt.legend()
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
plt.title('Probability density funtion',fontsize=7.5)
plt.xlabel('x (distance)',fontsize=5)
plt.ylabel('P(x)',fontsize=5)
plt.legend(loc='upper right', scatterpoints = 1,prop={'size':5})
plt.xlim([0,1])
plt.grid()

plt.subplot(2,2,2)
for u,c in zip(coverage,colors):
    y=1-np.exp(-u*x)
    plt.plot(x,y,'-',linewidth=2,alpha=1,color=c,label='$\lambda$ =%.3f'%u)
    plt.fill_between(x,y,color=c,alpha = .25)
plt.legend()
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
plt.title('Cumulative distribution funtion',fontsize=7.5)
plt.xlabel('x (distance)',fontsize=5)
plt.ylabel('P(event <= x)',fontsize=5)
plt.legend(loc='upper right', scatterpoints = 1,prop={'size':5})
plt.grid()

plt.tight_layout()	
```

In the case of sequencing, we can see from the PDF that increacing $\lambda$ drives the PDF closer to the Y-axis. This simply
means that more of the AUC is traversed for smaller increments in X. This can also be seen in the CDF, as the probability of
observing an event for small increments in X rapidly increaces in proportion to $\lambda$. In short, the distance between reads, assuming that read "arrivals" on the genome is an exponential random variable, will be shorter as coverage increaces. 

#### Continuous distributions: Normal -

The probability density funtion for the normal distribution is: 

$$ P(x) = \frac{1}{\sigma \sqrt{2\pi}} exp(-\frac{(x-\mu)^{2}}{2\sigma^2}) $$

Note that the standard z-score is the number of standard deviations away from the mean:

$$ Z=\frac{(x-\mu)}{\sigma} $$

```python
# Specify parameters for normal distribution
colors = matplotlib.rcParams['axes.color_cycle']
stdVal=3

plt.subplot(2,2,1)
mean = [5,10,20]
for meanVal,c in zip(mean,colors):
    #  Generates a normal continuous random variable with specified mean (loc), std (scale) 
    rv = stats.norm(loc=meanVal,scale=stdVal)   
    x = np.arange(.1*meanVal,meanVal*10,1)
    plt.plot(x,rv.pdf(x),lw=2,linewidth=2,alpha=1,color=c,label='$\mu$ =%.3f'%meanVal)
    plt.fill_between(x,rv.pdf(x),color=c,alpha = .25)
plt.legend()
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
plt.xlim([0,np.max(mean)*1.5])
plt.title('Probability density funtion',fontsize=7.5)
plt.xlabel('Random variable (x)',fontsize=5)
plt.ylabel('P(x)',fontsize=5)
plt.legend(loc='upper right', scatterpoints = 1,prop={'size':5})
plt.grid()

plt.subplot(2,2,2)
for meanVal,c in zip(mean,colors):
    rv = stats.norm(loc=meanVal,scale=stdVal)   
    x = np.arange(.1*meanVal,meanVal*10,1)
    plt.plot(x,rv.cdf(x),lw=2,linewidth=2,alpha=1,color=c,label='$\mu$ =%.3f'%meanVal)
    plt.fill_between(x,rv.cdf(x),color=c,alpha = .25)
plt.legend()
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
plt.xlim([0,np.max(mean)*1.5])
plt.title('Cumulative distribution funtion',fontsize=7.5)
plt.xlabel('Random variable (x)',fontsize=5)
plt.ylabel('P(event <= x)',fontsize=5)
plt.legend(loc='upper right', scatterpoints = 1,prop={'size':5})
plt.grid()

plt.tight_layout()
```

Evaluating the CDF on intervals around the mean value will clearly show that (1) About 68% of values drawn from a normal distribution are within one standard deviation $\sigma$ away from the mean. (2) About 95% of the values lie within two standard deviations. (3) About 99.7% are within three standard deviations. 

### Inferential Statistics and Hypothesis Tests

#### Central Limit Theorm -

Given any distrbution, take samples of size $s$ from it and plot the frequency of sample means. As $s$ increases, (1) the histogram of sample means will become more *normally distributed* with $\mu_{sample} = \mu_{population}$, meaning that the mean of the sampling distribution will be equal to the mean of the population distribution. (2) The standard deviation gets smaller. Knowing $s$ and $\sigma_{population}$, we can predict the standard deviation of the sampling distribution of the sample mean as $\frac{\sigma_{population}}{\sqrt{N}}$.

```python
ndata = 1e5 # 100k random datapoints between 0 and 1
lambda_ = 4.5 # expected value of 4.5
data = np.random.poisson(lam=lambda_, size=ndata) # draw Poisson random variables
plt.subplot(2,2,1)
colors = matplotlib.rcParams['axes.color_cycle']
sampleSize=[5,10,50,100] # sample sizes
for ss,c in zip(sampleSize,colors):
    sumData=np.mean(data.reshape(int(ndata/ss),ss),axis=1) # sample means
    hist,bins=np.histogram(sumData, bins=np.linspace(0,np.max(sumData),50))
    center=(bins[:-1] + bins[1:])/2
    plt.plot(center,np.array(hist),linewidth=2,color=c,label='Sample size %s'%ss)
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
plt.xlabel('Sample mean',fontsize=5)
plt.ylabel('Counts per bin',fontsize=5)
plt.title('Sampling distribution of sample mean: \n Histogram of sample means',fontsize=7.5)
plt.legend(loc='upper right', scatterpoints = 1,prop={'size':5})
plt.grid()
plt.tight_layout()	
```

In many cases in inferetial statistics, we want the evaluate the probability of obtaining a test statistic (e.g., a sample mean) assuming a population (distribution) from which the sample was picked. As explained above, if we pick samples of size $s$ from a population, the sampling distribution of sample means will approach $\mu_{population}$ as $s$ increaces, and the standard deviation of the sampling distribution of the sample mean is $\frac{\sigma_{population}}{\sqrt{s}}$. Thus, is $s$ or $N$ (below) is sufficiently large, the mean $\mu$ of the sample distribution is equal to that of the population and $\frac{\sigma_{population}}{\sqrt{N}}$. In many cases, we don't know $\sigma_{population}$ and assume that $\sigma_{population}=\sigma_{sample}$ (often if $N>30$). 

```python
# Run a single test
populationMean=1.2
populationStd=0.05
#  Generate a normal continuous random variable with specified mean (loc), std (scale) 
plt.subplot(2,2,1)
rv_pop = stats.norm(loc=populationMean,scale=populationStd)   
x = np.arange(populationMean-3*populationStd,populationMean+3*populationStd,.01)
plt.plot(x,rv_pop.pdf(x),lw=2,alpha=1,color='blue',label='Population. $\mu$=%s, $\sigma$=%s'%(populationMean,populationStd))
limits=rv_pop.interval(0.95)
plt.axvline(limits[0], color='blue',linestyle='solid',lw=1) 
plt.axvline(limits[1], color='blue',linestyle='solid',lw=1) 
# Pick a sample from the distribution
N=100
sampleMean=populationMean
sampleStd=populationStd/(N**(0.5))
rv = stats.norm(loc=sampleMean,scale=sampleStd)   
plt.plot(x,rv.pdf(x),lw=2,alpha=1,color='red',label='Sample. $\mu$=%s, $\sigma$=%s'%(sampleMean,sampleStd))
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
plt.title('Population versus sample distribution',fontsize=7.5)
plt.xlabel('Random variable (x)',fontsize=5)
plt.ylabel('P(x)',fontsize=5)
plt.legend(loc='upper left', scatterpoints = 1,prop={'size':5})
plt.grid()
```

For the population, there is a 95% probability of obtaining a value between the interval bounds. As shown above, the PDF for the sampling distribution of sample means has the same mean as the PDF for the population, but a tigher distribution given by $\frac{\sigma_{sample}}{\sqrt{N}}$ where we assume $\sigma_{sample}=\sigma_{population}$. If we pick a sample from the population, we see that the $\sigma$ for the sample (standard error of the sampling mean) is smaller than the population, producing a sharper curve. Given a sample of size $N=100$ with mean $\mu=1.05$ and $\sigma=0.5$, we can now determine whether this is a significant deviation from the population mean $\mu_{0}=1.2$. We can assume that the null hypothesis (no difference) is true, compute the sampling distribution, and determine the probability of obtaining the Z-statistic associated with $\mu=1.05$ (which assumes $\sigma_{sample}=\sigma_{population}$ for $N>30$): 

$$ Z=\frac{(\mu_{0}-\mu)}{\frac{\sigma_{sample}}{\sqrt{N}}} $$

```python
# Typical P-value for rejecting the null hypothesis
pVal=0.05

# Run a single test
sampleMean=1.05
stdVal=0.05
meanVal=1.2
z=(1.2-sampleMean)/stdVal
limits=rv.interval(1-pVal)

plt.subplot(2,2,1)
#  Generate a normal continuous random variable with specified mean (loc), std (scale) 
rv = stats.norm(loc=meanVal,scale=stdVal)   
x = np.arange(meanVal-5*stdVal,meanVal+5*stdVal,.01)
plt.plot(x,rv.pdf(x),lw=2,alpha=1,color='black',label='$\mu$ =%.3f'%meanVal)
plt.fill_between(x, rv.pdf(x), where=(x < limits[0]),alpha=1,color='black')
plt.fill_between(x, rv.pdf(x), where=(x > limits[1]),alpha=1,color='black')
plt.legend()
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
plt.xlim([meanVal-4*stdVal,meanVal+4*stdVal])
plt.title('Probability density funtion',fontsize=7.5)
plt.xlabel('Random variable (x)',fontsize=5)
plt.ylabel('P(x)',fontsize=5)
plt.legend(loc='upper right', scatterpoints = 1,prop={'size':5})
plt.axvline(sampleMean, color='red',linestyle='solid',lw=4) 
plt.grid()
```

If N<30, then we can instead use a T-statistic, which uses a T-table (rather than Z-table) and T-score (rather than Z-score).

#### Tests on two samples - 

Given two population that we want to compare, we take a sample from each and compute the means. The difference of the means is a random variable, Z. The mean of the distribution of Z is $\mu_{x-y}=\mu_x - \mu_y$ and the variance of Z is $\sigma_{x-y}=\sigma_x + \sigma_y$. Assuming normality, we can perform hypothesis tests on this difference using Z or T-tables, as discussed. If non-normal, then we need to use a non-parametric test such as the Mann-Whitney-Wilcoxon (MWW) RankSum test. This is a simple algorithm that ranks the data, counts the number of observations in one group that have a smaller rank than the other, and sums these to produce U, a test statistic whose distribution under the null hypothesis is known. It is often used to evaluate whether a particular population tends to have larger values than the other.

```python
plt.subplot(2,2,2)
ndata = 100
data = np.random.random(ndata)
twoGroups=data.reshape(int(ndata/2),2)
scale=1.3
twoGroups[:,0]=twoGroups[:,0]*scale
z_stat, p_val = stats.ranksums(twoGroups[:,0], twoGroups[:,1])  
print "P-value fro Wilcoxon test is: %0.3f"%p_val
plt.boxplot(twoGroups)
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
plt.ylabel('Value',fontsize=5)
plt.title('Wilcoxon RankSum test on two samples',fontsize=7.5)
plt.grid()
plt.tight_layout()	
```

#### Multiple hypothesis testing - 

Assuming the null hypothesis is true, we generate a sampling disrubtion with mean $\mu$ equal to that of the population and standard error of the sampling mean. The observed sample mean has a low probability (or P-value ~ 0.03 as it has a Z-score ~ 3) of occouring. This P-value is below the threshold of 0.05, so we reject the null hypothesis. However, suppose we obtain many samples and perform this same test. As an exercise, let's pick samples from the normal distribution itself. We can see that as the sample number increaces, the number values that exceed the naive P-value threshold inceace. A more stringent metric is the Bonferroni P-value:

$$ P_{Bon}=\frac{P}{n} $$

```python
# Illustrate the multiple hypothesis testing fallacy
plt.subplot(2,2,1)
numVals=[10,100,1000,10000]
outliers=[]
outliers_correctedP=[]
meanVal=1.2
stdVal=0.05
pVal=0.05
colors = matplotlib.rcParams['axes.color_cycle']
rv = stats.norm(loc=meanVal,scale=stdVal)  
for num,c in zip(numVals,colors):
    # Generate random values from the distribution
    randomValues=stdVal*np.random.randn(num)+meanVal
    # Number for each point
    pointNum=np.linspace(1,num,num)
    # Scatter
    plt.scatter(np.array(pointNum),np.array(randomValues),color=c,alpha=0.25,label='Trials=%i'%num)
    # Record number of values > P threshold
    limits=rv.interval(1-pVal)
    high_Outliers=randomValues[randomValues>limits[1]].size
    outliers=outliers+[high_Outliers]
    # Correct the P value (Bonferroni) and record values again
    pVal_correct=pVal/num
    limits_stringent=rv.interval(1-pVal_correct)
    high_Outliers=randomValues[randomValues>limits_stringent[1]].size
    outliers_correctedP=outliers_correctedP+[high_Outliers]
# Limits that contain 95% of the data
limits=rv.interval(1-pVal)
plt.fill_between(pointNum,limits[0],limits[1],facecolor='black', alpha=0.5, label='P-value')
plt.legend(loc='center', scatterpoints = 1,prop={'size':5})
plt.xlim([1,num])
plt.title('Sampled points from normal \n P-value threshold %s'%pVal,fontsize=5)
plt.xlabel('Point',fontsize=5)
plt.ylabel('Value',fontsize=5)
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
plt.grid()

# Bonferroni correction
plt.subplot(2,2,2)
plt.plot(outliers,'-o',alpha=0.5,color='red',lw=2,label='Non-corrected P-value')
plt.plot(outliers_correctedP,'-o',alpha=0.5,color='blue',lw=2,label='Bonferroni corrected P-value')
plt.title('Number points from apparently come \n from different distribution',fontsize=5)
plt.legend(loc='upper left', scatterpoints = 1,prop={'size':5})
ind=np.arange(len(numVals))
locs,labels=plt.xticks(ind,numVals)
plt.setp(labels, rotation=90, fontsize=5)
plt.yticks(fontsize=5)
plt.ylim([-0.1*plt.ylim()[1],plt.ylim()[1]*1.1])
plt.xlabel('Points sampled',fontsize=5)
plt.ylabel('Value',fontsize=5)
plt.grid()

plt.tight_layout()
```

As the numnber of values we randomly pick from the distribution, some are outliers and have a low P-value. As we can see, they fall outside of the blacklisted region that holds 95% of the data. Therefore, these points would pass the P-value threshold and be wrongly attributed from a different distribution (false positives). We can use a Bonferroni correct P-value threshold to account for this: as we can see, the false-positives do not pass this more stringent P-value filter. Because Bonforroni is highly conservative, we can also use the Benjami-Hochberg correction. In this case, i is the rank order of the hypothesis, m is the total number of hypotheses, and $\alpha$ is the cutoff (e.g., 0.05). This algorithm will find the highest rank, $i$, for which this condition holds and reject the null hypothesis (e.g., that the data come from the same distribution) for all rank lower than this.

$$ P_i < \frac{i}{m}\alpha $$

```python
# General structure of FDR confusion matrix
index=["+","-",'Sum']
columns=["+","-",'Sum']
matrix=pd.DataFrame(index=index,columns=columns)
matrix.loc["+","+"]='S:True Positive'
matrix.loc["+","-"]='V:False Discovery'
matrix.loc["+","Sum"]='All Discoveries'
matrix.loc["-","+"]='T:False Negative'
matrix.loc["-","-"]='U:True Negative'
matrix.loc["-","Sum"]='m-R'
matrix.loc["Sum","+"]='m-m_o'
matrix.loc["Sum","-"]='m_o:Negatives'
matrix.loc["Sum","Sum"]='m:All hypotheses tested'
matrix
```

In this case, the FDR level is:

$$ FDR = \frac{V}{D} $$

Using this correction, the $ FDR < \frac{m_o}{M}\alpha $. 

```python
# FDR correction
from statsmodels.stats.multitest import multipletests
# Generate random values from the distribution
numVals=10000
randomValues=stdVal*np.random.randn(num)+meanVal
# Z-score (all absolute value)
zScores=np.abs(meanVal-randomValues)/stdVal
# Re-compute (positive) value
posVals=meanVal+zScores*stdVal
# Percentage of data < the given value
pValues=1-rv.cdf(posVals)
# Benjamini-Hochberg false discovery rates
pm = multipletests(pValues, method="fdr_bh")
# Adjusted p-values are in the first position of the returned value
apv = pm[1]
ii = np.argsort(pValues)
plt.subplot(2,2,3)
plt.plot(pValues[ii],apv[ii],lw=2,label='Corrected P-values')
plt.xlim([-0.1*plt.xlim()[1],1.1*plt.xlim()[1]])
plt.ylim([-0.1*plt.ylim()[1],1.1*plt.ylim()[1]])
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
plt.title('FDR: P-value correction',fontsize=5)
plt.xlabel('Inital P-value',fontsize=5)
plt.ylabel('FDR corrected P-value',fontsize=5)
plt.legend(loc='lower right', scatterpoints = 1,prop={'size':5})
plt.grid()

pValues_sort=pValues[np.argsort(pValues)]
plt.subplot(2,2,4)
plt.plot(pValues_sort,c='blue',label='Uncorrected P-values')
# Threshold value
threshold=[pVal*i/numVals for i in range(1,len(pValues_sort))]
plt.plot(threshold,c='red',label='Cutoff for p=%.3f'%pVal)                   
plt.yticks(fontsize=5)
plt.title('FDR: Define corrected cutoff',fontsize=5)
plt.xlabel('Rank order',fontsize=5)
plt.ylabel('P-values',fontsize=5)
plt.legend(loc='upper left', scatterpoints = 1,prop={'size':5})
plt.grid()
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)

plt.tight_layout()
```

### Clustering (Unsupervised Machine Learning)

A way to group of similar things. A common application is gene expression, as the expression of genes are often queried across conditions (pertubations to the cell, cell types, cancer samples, time series data). Microarrays first addressed this, taking advantage of the fact that 'Watson' stand DNA wil transcribe 'Crick' strand RNA, which is then converted back to 'Watson' strand cDNA and hybridized to 'Crick' bait on a slide. Fold changes in detected levels of a given gene can be due to (1) up-regulation or (2) copy number changes (often in the case of Cancer). Therefore, the input to the clustering algorithm is (1) vectors of information and (2) a distance measure. The output is groupings of "similar" vectors (by the distance measure specified). ("Heristic" vs. polynomial?)

#### Distances - 

(1) Euclidian distance (between vectors - two lists of numbers): 
$$ \sqrt{\sum (x_i - y_i)^2} $$ 

(2) Correlation (of two lists of numbers):
$$ \frac{cov(x,y)}{\sigma_x \sigma_y} $$ 

(3) Manhattan (k=1) and Cheychev (k=inf) and k-th root:
$$ \sqrt{\sum (x_i - y_i)^k} $$ 

(4) *** Angle ***
$$ acos \frac{abs(x \cdot y)}{abs(x) abs(y)} $$ 

#### Evaluate clusters - 

(1) The intra-cluster distance (average or max distance between elements within a cluster) 
(2) The inter-cluster distance (average or max or min distance to another cluster). These parameters can indicate whether there are indeed clusters within the data. Silhouette measurment is a metric to evaluate the quality of clustering: a(i) = average distance of point i to all other points in the cluster, b(i) is the lowest average distance to neighboring cluster. Perform for all points.

$$ S(i) = \frac{b(i)-a(b)}{max(a(i),b(i))} , S(i) \rightarrow 1 , b(i) > a(i) $$

#### K means clustering - 

Algorithm: (1) Pick K (number of centers), (2) Generate K random points (centers), (3) Compute distance of each data point to each center, (4) Assign datapoint to closest center,  (5) Compute new center location as the average of all points assigned to center. (6) Loop until centers no long move. (6) Compute S values for all points to evaluate assinged clusters.  

```python
import sklearn.datasets as datasets
# Make data with four clusters (X are data, Y are centers)
X,Y = datasets.make_blobs(centers=4, cluster_std=0.5, random_state=0)
plt.subplot(2,2,1)
plt.jet() # Set the color map.
plt.scatter(X[:,0], X[:,1],c=Y)
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
plt.grid()
plt.title('Pre-clustered data',fontsize=7.5)

plt.subplot(2,2,2)
from sklearn.cluster import KMeans
kmeans = KMeans(4, random_state=8) # Random state is the generator used to initialize the centers
Y_hat = kmeans.fit(X).labels_
plt.scatter(X[:,0], X[:,1], c=Y_hat, alpha=0.4)
mu = kmeans.cluster_centers_
plt.scatter(mu[:,0], mu[:,1], s=100, c=np.unique(Y_hat))
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
plt.grid()
plt.title('Assign clusters with kMeans',fontsize=7.5)
```

#### Hierarchical clustering - 

$N^2$ Algorithm: (1) Each data vector is a node. (2) Compute distance of all nodes to one another: the matrix is symmetric (order $\frac{N^2-N}{2}$), as we can subtract each measurment of a node to itself (N) and divide by two because of symmetry (3) Merge the two "closest" nodes into a single node, based upon the criteria of interet (Single linkage: Minimum distance, Complete linkage: Maximum distance, Average linkage: Average distance). (4) Compute distance of new node to all. (5) Continue until all data is in one node. (6) "Prune" the heirarchy into clusters of interest.

#### Cluster enrichment - 

(1) Gene Ontology: Formal ontology for describing which processes / tissues specific genes are involoved in by (a) funtion (binding, enzyme, molecular level description), (b) process (visual, sesing), (c) component (nucleous). (2) Enrichment: Is a GO code enriched in a cluster? Given N total genes and M gene annotated with $GO_{i}$ and n genes in cluster x annotated $GO_{i}$, enrhchment may be calculated using the hypergeometric distribution in order to indicate the probability of observing $x$ genes in the cluster by chance:

$$ p = {M \choose x} {N-M \choose n-x} \frac{1}{N \choose n}   $$

Need to be careful here. Examine values of the Binomal coef. If $k=N$, then ${N \choose k}$ = 1, as there is only one way to arrange all k events. If $k=1$, then there are $N$ ways to arrage the single event. The the number of combinations increaces between those edge cases, which is useful to keep in mind when examining how enrichement works.

```python
# Binomal coef
def binomialCoef(N,k):
    val=math.factorial(N)/(math.factorial(N-k)*math.factorial(k))
    return val
N=5
ks=range(1,6,1)
out=[binomialCoef(N,k) for k in ks]
plt.subplot(2,2,1)
plt.plot(ks,out,lw=2)
plt.ylim([0,plt.ylim()[1]*1.1])
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
plt.xlabel('k',fontsize=5)
plt.ylabel('Combinations given N=%i'%N,fontsize=5)
plt.title('Values of Binomal Coef',fontsize=7.5)
plt.grid()
```

For high enrichment we want, we want a low probability of observing x genes in the cluster this by chance: (1) $N \choose n$ needs to be big. For that, clusters should not be huge, as $n$ should not approach $N$. If clusters are small (small $n$), then $N$ should be large. The optimal case is a large number of combinations given by moderate n relative to N. (2) ${M \choose x}$ should be small. This is clear: optimally, $x$ = $M$, meaning that all M  $GO_{i}$ genes are in the cluster. (3) The final term is the most conflusing: we want ${N-M \choose n-x}$ to be small. But if the $N-M ~ n-x$, then all "background genes" would be in the cluster, making the cluster size on par with $N$. This would drive up the denominator. Therefore, I think the optimal case is a very low value of $n-x$, meaning that the cluster has few background genes, which is intutive, a reasonably large N, which will drive up the denominator, and a large $n$. This basically says that there is a high background to choose from but the cluster contains little of it, and that the cluster contains a high fraction of the total $M$ $GO_{i}$ associated genes.

 $$ p = \sum \limits_{j=x}^n {M \choose x} {N-M \choose n-x} \frac{1}{N \choose n} = 1 -  \sum \limits_{j=0}^x {M \choose x} {N-M \choose n-x} \frac{1}{N \choose n} $$

We can compute the proability of seeing x $GO_{i}$ genes in the cluster as 1 - the probability of seeing at most x genes.

### Classification (Supervised Machine Learning) 

Classification starts with labeled data (supervised vs. clustering with unlabeled data) and finds a funtion / model that predicts (or classifies) a label for new points.

#### kNN - 

We can start by using a binary kNN classifer. Given (1) labeled vectors, (2) distance measure, (3) k neighbors to evaluate, and (4) unlabeled data to evaluate, our funtion will output a labels for each unlabeled data point. To do this, it finds the k nearest (labeled) data points and uses a vote to assign the label. We can also overlay weighting (e.g., based upon distance) such that closer labeled data to the query have a strong influence on the assignment than farther labeled data.

```python
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets

# Assign neighbors
n_neighbors = 15
# Import some data to play with
iris = datasets.load_iris()
X = iris.data[:, :2]  # Take the first two features (Sepal Length, Sepal Width)
y = iris.target # 3 different types of irises
# Create color maps
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
# Param for classifier
weights='distance'
# Create an instance of Neighbours Classifier and fit the data
clf = neighbors.KNeighborsClassifier(n_neighbors, weights=weights)
# Train the classifer on the data
clf.fit(X, y) 
# Make a grid of points that spand the length, width dims
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
h = .02  # step size in the mesh
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),np.arange(y_min, y_max, h))
# Label each new point based upon trained classifier
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
# Assign a color to each point in the mesh
Z = Z.reshape(xx.shape)
plt.subplot(2,2,1)
# Plot decision boundary
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
# Plot training data
pl.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold)
pl.xlim(xx.min(), xx.max())
pl.ylim(yy.min(), yy.max())
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
plt.xlabel('Feature 1: Sepal Length',fontsize=5)
plt.ylabel('Feature 2: Sepal Width',fontsize=5)
pl.title("Decision boundary and training data for k = %i" % n_neighbors,fontsize=7.5)
plt.grid()
plt.tight_layout()
```

#### Linear model - 

Basic linear regression uses a model to descibe data, with standard error indicating the degree of variation from the data:

$$ \sum (y_{i}-y_{model})^2 $$

The total variation in Y (related to the variance) is:

$$ \frac{\sum (y_{i}-y_{mean})^2}{N} $$

Thus, the percentage of variation that is not described by the line is:

$$ \frac{\sum (y_{i}-y_{model})^2}{\sum (y_{i}-y_{mean})^2} $$

1 minus this value is the variation described by the line, otherwise known as the coef of determination or $r^{2}$.

```python
from sklearn.utils import shuffle
# Load the diabetes dataset (10 physiological variables on 442 patients)
diabetes = datasets.load_diabetes()
X = diabetes.data
y = diabetes.target
# Shuffle
X, y = shuffle(X, y, random_state=1)
# Use only one column from data (one parameter)
X = X[:, 2:3] 
# Split into training and test data
train_set_size = 250
X_train = X[:train_set_size]  # selects first 250 rows (examples) for train set
X_test = X[train_set_size:]   # selects from row 250 until the last one for test set
y_train = y[:train_set_size]   # selects first 250 rows (targets) for train set
y_test = y[train_set_size:]    # selects from row 250 until the last one for test set
# Plot raw data
plt.subplot(2,2,1)
plt.scatter(X_train, y_train)
plt.scatter(X_test, y_test)
plt.xlabel('Data | Measured value',fontsize=5)
plt.ylabel('Target | Patient',fontsize=5)
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
pl.title("Raw data",fontsize=7.5)
plt.grid()

# Linear regression object
from sklearn import linear_model
regr = linear_model.LinearRegression()
# Fit training data
regr.fit(X_train, y_train)
plt.subplot(2,2,2)
plt.scatter(X_train, y_train,  color='blue',alpha=0.5,label='Training data')
plt.scatter(X_test, y_test,  color='red',alpha=0.5,label='Test data')
plt.plot(X_train, regr.predict(X_train), color='blue', linewidth=2)
plt.legend(loc='upper left', scatterpoints = 1,prop={'size':5})
plt.xlabel('Data',fontsize=5)
plt.ylabel('Target',fontsize=5)
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
pl.title("Model fit to data \n Coef=%.3f \n Intercept=%.3f \n r^2=%.3f"%(regr.coef_,regr.intercept_,regr.score(X_train, y_train)),fontsize=7.5)
plt.grid()

# Standard error
print("Training error: ", np.sum((regr.predict(X_train) - y_train) ** 2))
print("Test error: ", np.sum((regr.predict(X_test) - y_test) ** 2))

# Meam squared error
print("Training error: ", np.mean((regr.predict(X_train) - y_train) ** 2))
print("Test error: ", np.mean((regr.predict(X_test) - y_test) ** 2))

plt.tight_layout()
```

#### Decision trees and random forrests - 

Decision trees draw lines through the feature space to classify the data, creating a set of boolean rules that apply to the features such as $f_{1} > k_{1}$ and so forth. #### Naive Bayes - 

Frequentist probability relies on the long-run frequency of events (e.g., the probability of plane accidents under a frequentist philosophy is interpreted as the long-term frequency of plane accidents). Bayesian probability assigns a belief, or prior probability, to an unlabeled point $P(H)$. Thus, we have to state our prior probability that a point has a specific label.
We also must state the probabilty of observing the data point $P(D)$ and the probability of observing the point if the hypothesis is true is $P(D|H)$. From this we can estimate the probability that the point is labeled given the data $P(H|D)$, the posterial probability. We use the chain rule to de-compose D into its features.

$$ P(H|D) = \frac{P(D|H)P(H)}{P(D)} \approx P(H) P(D_1 | H) P(D_2 | D_1 H) P(D_3 | D_2 . . . ) \approx H \prod_{i=1}^n P(D_i|H) $$

We make the naive assumption that all the features are independent, and thus each is $P(D_i|H)$ without contribution from other features. Givem a point and hypothesis that it is blue, we compute the probability of observing a blue point along the values of each feature (by binning it). This gives us the $P(D_i|H)$ value, and we multiply across all features.

#### Cross - validation and ROC curves - 

Sensitivity is the ability to pick up all of the positives (recall) wheras positive predictive value indicates our ability to detect all of the positives accuratly. Perfect sensitivity (always calling POSITIVE) will trade-off poor sensitivity due to the high number of false positives, whereas high specifity will suffer from poor sensitivity (no true POSITIVES called).

$$ Sensitivity = \frac{TP}{TP+FN} $$

$$ Specificity = \frac{TN}{TN+FP} $$

$$ PPV = \frac{TP}{TP+FP} $$

$$ NPV = \frac{TN}{TN+FN} $$

ROC curves can be used summarize the performance of a classifier by modulating a parameter that will trade-off between false positives and false negatives, plotting false positive rate (x) versus true positive rate (y). If all TPs can be called without any FPs, then there is simply a veritcal line at FPR=0 (as shown in the simple case below with logistic classification). If more FPs must be incurred to get all the TPs, then the line is staggered (second example below).

$$ FPR = 1-Specificity = 1-\frac{TN}{TN+FP} $$

$$ TPR = Sensitivity = \frac{TP}{TP+FN} $$

These are very useful for PCA, a way to find the directions (components) that maximize the variance in our dataset. 

```python
# Generate 40 3-dimensional samples randomly drawn from a multivariate Gaussian distribution.
np.random.seed(234234782384239784) # random seed for consistency
mu_vec1 = np.array([0,0,0])
cov_mat1 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class1_sample = np.random.multivariate_normal(mu_vec1, cov_mat1, 20).T
assert class1_sample.shape == (3,20), "The matrix has now the dimensions 3x20"
mu_vec2 = np.array([1,1,1])
cov_mat2 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class2_sample = np.random.multivariate_normal(mu_vec2, cov_mat2, 20).T
assert class1_sample.shape == (3,20), "The matrix has now the dimensions 3x20"
all_samples = np.concatenate((class1_sample, class2_sample), axis=1)
assert all_samples.shape == (3,40), "The matrix has now the dimensions 3x40"
# Means
mean_x = np.mean(all_samples[0,:])
mean_y = np.mean(all_samples[1,:])
mean_z = np.mean(all_samples[2,:])
mean_vector = np.array([[mean_x],[mean_y],[mean_z]])
print('Mean Vector: \n', mean_vector)
```

```python
# Compute the covariance matrix
cov_mat = np.cov([all_samples[0,:],all_samples[1,:],all_samples[2,:]])
print('Covariance Matrix:\n', cov_mat)
# Eigenvectors and eigenvalues for the from the covariance matrix
eig_val_cov, eig_vec_cov = np.linalg.eig(cov_mat)
# 1) First eigenvector is the direction in which the data varies the most
# 2) Second eigenvector is the direction of greatest variance among those that are orthogonal (perpendicular) to the first eigenvector
# 3) The eigenvalues are the length of the arrows.
for i in range(len(eig_vec_cov)):
    eigvec_cov = eig_vec_cov[:,i].reshape(1,3).T
    print('Eigenvector {}: \n{}'.format(i+1, eigvec_cov))
    print('Eigenvalue {} from covariance matrix: {}'.format(i+1, eig_val_cov[i]))
    print(40 * '-')
```

If data are standardized such that each sample is centered to zero average expression level, the principal components are normalized eigenvectors of the covariance matrix of the samples and ordered according to how much of the variation present in the data they contain. 

The first eigenvector points (from the mean of the data) in the direction in which the data varies the most in Euclidean space.

```python
%pylab inline
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d import proj3d
from matplotlib.patches import FancyArrowPatch

class Arrow3D(FancyArrowPatch):
    def __init__(self, xs, ys, zs, *args, **kwargs):
        FancyArrowPatch.__init__(self, (0,0), (0,0), *args, **kwargs)
        self._verts3d = xs, ys, zs

    def draw(self, renderer):
        xs3d, ys3d, zs3d = self._verts3d
        xs, ys, zs = proj3d.proj_transform(xs3d, ys3d, zs3d, renderer.M)
        self.set_positions((xs[0],ys[0]),(xs[1],ys[1]))
        FancyArrowPatch.draw(self, renderer)

fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111, projection='3d')
ax.plot(all_samples[0,:], all_samples[1,:], all_samples[2,:], 'o', markersize=8, color='green', alpha=0.2)
ax.plot([mean_x], [mean_y], [mean_z], 'o', markersize=10, color='red', alpha=0.5)
for v in eig_vec_cov.T:
    a = Arrow3D([mean_x, v[0]], [mean_y, v[1]], [mean_z, v[2]], mutation_scale=20, lw=3, arrowstyle="-|>", color="r")
    ax.add_artist(a)
ax.set_xlabel('Feature_1',fontsize=7.5)
ax.set_ylabel('Feature_2',fontsize=7.5)
ax.set_zlabel('Feature_3',fontsize=7.5)
```

Eigenvectors will form the axes of this new feature subspace. The eigenvectors with the lowest eigenvalues bear the least information about the distribution of the data, and those are the ones we want to drop. The common approach is to rank the eigenvectors from highest to lowest corresponding eigenvalue and choose the top k eigenvectors.

```python
# Make a list of (eigenvalue, eigenvector) tuples
eig_pairs = [(np.abs(eig_val_cov[i]), eig_vec_cov[:,i]) for i in range(len(eig_val_cov))]
# Sort the (eigenvalue, eigenvector) tuples from high to low
eig_pairs.sort()
eig_pairs.reverse()
# Visually confirm that the list is correctly sorted by decreasing eigenvalues
for i in eig_pairs:
    print(i[0])
# We are combining the two eigenvectors with the highest eigenvalues to construct
matrix_w = np.hstack((eig_pairs[0][1].reshape(3,1), eig_pairs[1][1].reshape(3,1)))
print('Matrix W:\n', matrix_w)
```

We transform our samples onto the new subspace via the equation $y=W^t \times x$.

```python
transformed = matrix_w.T.dot(all_samples)
assert transformed.shape == (2,40), "The matrix is not 2x40 dimensional."
plt.plot(transformed[0,0:20], transformed[1,0:20], 'o', markersize=7, color='blue', alpha=0.5, label='class1')
plt.plot(transformed[0,20:40], transformed[1,20:40], '^', markersize=7, color='red', alpha=0.5, label='class2')
plt.xlim([-4,4])
plt.ylim([-4,4])
plt.xlabel('PC_1',fontsize=7.5)
plt.ylabel('PC_2',fontsize=7.5)
plt.legend()
plt.title('Transformed samples with class labels',fontsize=10)
plt.grid()
```

The new axes represent a linear combination of the initial features: $PC_1 = \alpha Feature_1 + \beta Feature_2 + \gamma Feature_3$.

#### Global (Needleman-Wunsch) and Local (Smith-Waterman) alignment (Order $N^2$ algorithm) - 

A sequence alignment is a set of hypotheses about the equivalance (structural/funtional role) between two sequences. If can be used to (1) infer common descent, as difference in sequence can correlate with time since divergence or (2) infer similar biological structure / funtional roles. $\textbf{Aligned amino acids:}$ Identifical amino acids are often aligned (there are 20 AAs total), but non-identical AAs with similar funtional properties (e.g., same change) may also be aligned, as the funtion can be perform by numerous amino acids. $\textbf{Gaps:}$ Show the loss or gain of elements since the time of divergence between the sequences. $\textbf{The gold standard:}$ Structural alignment, manually aligned based upon biophysical knowledge of similar structural / funtional role. $\textbf{Match matrix}$: Residues have diverse biophysical properties and diverse scores. The PAM250 matrix is popular, and is approx how frequently AA substitutions are observed in organisms that diverse by 50 generations. It is shown below, with p(i) as p(AAi) in DB, q(i,j) as the probability of alignment, and $\lambda$ as a scale factor.

$$ S(i,j) = \frac{ln(q(i,j)/p(i)p(j)}{\lambda}$$ 

Blosum50 assumes a 3D shared structure for a block of sequence and examines empirical substiutions. $\textbf{Gap penalty}$: Can be a simple $-gd$, though better results are obtained using the Affine gap. This is has an opening penality and an extension penality. $\textbf{Gloal alignment algorithm}$: Gaps make the problem hard, as they can be introduced in either sequence at any position for any length. Addressed using dynamic programming. Find optimal alignment that ends with A[M]-B[N] or A[M]-_ or _-B[B]. Do this by filling the $\textbf{Score matrix}$, obtain score at final elemnts F[A,B], and trace back to obtain the full alignment. Note that three matricies are needed when using the Affine gap. $\textbf{Local alignment algorithm}$: Adjust the score matrix so that it always contains values $> 0$ such that the penality for poor score are not carried forward, but matrch matrix has negative values, and traceback can start anywhere (specifically, the largest value in the matrix, as we want to see the path used to get there).

#### Similarity versus Homology - 

The alignment score is similarity. Homology indicates common descent (as homologues play similar biological roles in different systems). We cannot infer homology; this is determined from speciation. But, similarity is evidance for homology (common descent). Another explaination is convergent evolution (e.g., subject to a similar pressure). $\textbf{Orthology}$ A speciation event leads to B and C, which play the same structural/funtional role (thus B and C are orthologs). $\textbf{Paralog}$ Assume C' splits from C, as C' evolves a new funtion. Therefore similarity of funtion define orthologs whereas paralogs can have divergent funtion, yet all have common ancestors.

So, we can hash the Query into unique $kTup$ values and for each $kTup$ extract all addresses in the database. This return a bitvector for each $kTup$ (with offset based upon the order in the query). BLAST uses a theory for score significance developed by Sam Karlin. BLAST defines the MSP, maximal segment pair. The highest scoring pair of identical length segments in random sequences of length M, N. As M, N get longer, the expected length will grow. This will also depend upon the matrch matrix, as this indicate the scores for base correspondences. Thus, knowing $N$ (size of database), $M$ (size of query), and the alignment score $S$, then the expected number of sequences scoring better than S:

$$ ~ k N M e^{-\lambda S} $$

#### SNPs, Recombination, Linkage disequilibrium - 

Driven by random mutations in a (primordial) human, which then led to M ancestors with the SNP today such that $\frac{M}{M+P}$ is the current population (with P being the remaining). Given a locus (an area of interest, which can be a single base), alleles are the choices that are available at that locus. SNPs are usually bi-allelic. SNPs can have mutual information or correlation, which is known as linkage disequilibrium. In the creation of gametes, there is a re-assortment of DNA from both parents. The games can take different SNPs from either Mom or Dad's chromosomal copy. $\textbf{LD is a measure of SNP correlation}$ such that $LD=0$ means no correlation and $LD>0$ means some correlation. In a simple example, two parents can have the following sequence of SNPs across an allele of interest. In this case, the presence of SNP1 predicts 3.

An individual has two copies of a given chromosome (one from each parent). $$\\$$ 
Parent 1 chromosome: $$A-A-G$$
Patent 2 chromosome: $$G-A-T$$

For any two alleles $A_1$ and $B_1$:

$$ SNP_{A} \rightarrow A_{1} or A_{2} $$ 
$$ SNP_{B} \rightarrow B_{1} or B_{2} $$ 
$$ f(A_1 , B_1) $$
$$ D = f(A_1 , B_1) - f(A_1)f(B_1) $$

If independent, then:

$$ f(A_1 , B_1) = f(A_1)f(B_1) $$

It is also common to normalize D as $D_{max}$, which is the maximum value of D given the allele frequencies:

$$ D' = \frac{D}{D_{max}} $$
$$ D_{max}: $$ 
$$ If D < 0 = min (p(A_1)p(B_1) , p(A_2)p(B_2)) $$
$$ If D > 0 = min (p(A_1)p(B_2) , p(A_2)p(B_1)) $$

For any two loci, we can calculate the correlation:

$$ r = \frac{D}{\sqrt{p(A_1)p(A_2)p(B_1)p(B_2)}} $$

Most of the LD is due to new SNP events occouring on a background. In order to break LD, SNP events must be seperated by recombination. Thus, the limitations are that these (1) only measure 1M loci (leaving out rare variants), (2) are not universal to all ethnicities, and (3) only assays SNPs (rather than other variations, such as CNVs).
