```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt 
import pandas as pd
from pandas.io import sql
import sqlite3
import sys
import os
%matplotlib inline
matplotlib.rcParams['savefig.dpi'] = 2 * matplotlib.rcParams['savefig.dpi']
```

#### Read data - 

```python
# - Read data - 
outfilepathToSave = os.getcwd()+'/PR_sample_data.csv'
data_all_samples_bodysites = pd.read_csv(outfilepathToSave, header=[0, 1], index_col=[0, 1])
data_all_samples_bodysites.head(3)
```

#### Plot body site data - 

```python
# Choose specific body site to analyze
bodysite='Saliva'
```

```python
import brewer2mpl
global bmap 

def cm2inch(*tupl):
    inch = 2.54
    if type(tupl[0]) == tuple:
        return tuple(i/inch for i in tupl[0])
    else:
        return tuple(i/inch for i in tupl)

def barPlotData(dat_by_bodysite_norm_trimmed_simple,xlowlim):    
    print "Shape of data (bugs x samples) for the given bodysite ---"
    print dat_by_bodysite_norm_trimmed_simple.shape
    bugs=dat_by_bodysite_norm_trimmed_simple.index
    sample_names=dat_by_bodysite_norm_trimmed_simple.columns
    dataToPlot=np.array(dat_by_bodysite_norm_trimmed_simple)
    bottom=np.cumsum(dataToPlot,axis=0)
    ind=np.arange(dataToPlot.shape[1])
    plotBins=range(0,dataToPlot.shape[0])
    for j in range(0,dataToPlot.shape[0]):
        color = bmap[(j%colorMapMax)]
        if j==0:
            plotBins[j]=plt.bar(ind,dataToPlot[j,:],edgecolor='none',color=color,align='center',label=bugs[j]) 
        else:
            plotBins[j]=plt.bar(ind, dataToPlot[j,:],edgecolor='none',color=color,bottom=bottom[j-1,:],align='center',label=bugs[j]) 
    locs,labels=plt.xticks(ind,sample_names)
    plt.setp(labels, rotation=90, fontsize=5)
    locs,labels=plt.yticks()
    plt.setp(labels,fontsize=5)
    plt.legend(loc='upper left',numpoints=1,prop={'size':6})
    plt.ylim([0,1])
    plt.xlim([-11,plt.xlim()[1]])
    plt.xlim([xlowlim,dat_by_bodysite_norm_trimmed_simple.shape[1]])

# Returns an array: of bugs x sample_IDs
dat_by_bodysite=data_all_samples_bodysites[bodysite]
dat_by_bodysite.index=dat_by_bodysite.index.get_level_values('tax_name')
dat_by_bodysite.fillna(0,inplace=True)
dat_by_bodysite_norm=dat_by_bodysite.div(dat_by_bodysite.sum(axis=0),axis=1)

# View only bugs that are > cutoff in order to make viewing easier.
cutoff=0.075
dat_by_bodysite_norm_trimmed=dat_by_bodysite_norm[dat_by_bodysite_norm>cutoff]
dat_by_bodysite_norm_trimmed.fillna(0,inplace=True)
dat_by_bodysite_norm_trimmed_simple=dat_by_bodysite_norm_trimmed.loc[dat_by_bodysite_norm_trimmed.sum(axis=1)>0]
dat_by_bodysite_norm_trimmed_simple.loc['Other']=1-dat_by_bodysite_norm_trimmed.sum(axis=0)

# Plot
colorMapMax=12
bmap = brewer2mpl.get_map('Paired', 'Qualitative', colorMapMax).mpl_colors
fig=plt.figure(figsize=cm2inch(27.5,10))
xlowlim=-11
barPlotData(dat_by_bodysite_norm_trimmed_simple,xlowlim)
plt.title("Data for %s all pregnancy samples."%bodysite,fontsize=8)
```

#### Run PCA on all body sites - 

```python
# Prepare data
data_all_samples_bodysites_norm=data_all_samples_bodysites.div(data_all_samples_bodysites.sum(axis=0),axis=1)
data_all_samples_bodysites_norm_tp=data_all_samples_bodysites_norm.transpose()
body_site_index=data_all_samples_bodysites_norm_tp.index.get_level_values('BodySite')

# Run PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
siteData_r = pca.fit(data_all_samples_bodysites_norm_tp).transform(data_all_samples_bodysites_norm_tp)

# Plot
colors=['r','g','b','y','c','k']
uniq_sites=list(set(body_site_index))
fig=plt.figure(figsize=cm2inch(10,10))
for site,c in zip(uniq_sites,colors):
    plt.scatter(siteData_r[body_site_index==site, 0], siteData_r[body_site_index==site, 1], c=c, label=site, alpha=0.75)
plt.legend(loc='upper right',scatterpoints=1,prop={'size':4})
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
plt.xlabel('PC1',fontsize=5)
plt.ylabel('PC2',fontsize=5)
title='All body sites'
plt.title('PCA of %s data'%title,fontsize=5)
```

```python
def runLoadingPlot(target_name,pca):
    fig=plt.figure(figsize=cm2inch(12,10))
    loadings = pca.components_
    for i in range(2):
        storeLoad=pd.DataFrame()
        storeLoad['Names']=list(data_all_samples_bodysites_norm_tp.columns.get_level_values('tax_name'))
        storeLoad['Loadings']=loadings[i]
        storeLoad.set_index('Names',inplace=True,drop=True)
        storeLoad.sort('Loadings',ascending=False,inplace=True)
        ind=np.arange(storeLoad.shape[0])
        plt.subplot(2,1,i+1)
        plt.bar(ind,np.array(storeLoad['Loadings']))
        locs,labels=plt.xticks(ind,storeLoad.index)
        plt.setp(labels, rotation=90, fontsize=5)
        locs,labels=plt.yticks()
        plt.setp(labels,fontsize=5)
        plt.ylabel('Feature loading',fontsize=5)
        plt.title('PC%s for %s'%(i+1,target_name),fontsize=5)
    fig.tight_layout()
    
target_name='All body sites'
runLoadingPlot(target_name,pca)
```
