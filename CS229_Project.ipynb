```python
# - Autocorrelation plot - 

def k_autocorrelation(my_array, my_lag):
    ''' Kosh fxn '''
    # Params
    n = len(my_array)
    k = my_lag
    mu = my_array.mean()
    variance = my_array.var()
    # Normalize lengths and subtract mean
    my_array_lag = my_array[k:]
    my_array = my_array[:n-k]
    my_array_lag = my_array_lag - mu
    my_array = my_array - mu
    # Calculate R(k)
    R = (1/float(n-k)) * (1/float(variance)) * sum(my_array * my_array_lag)
    return R 

fig = plt.figure(figsize = (5, 2.5))
ax1 = fig.add_subplot(111)

# Autocorrelation using sm
# Docs = http://statsmodels.sourceforge.net/devel/generated/statsmodels.graphics.tsaplots.plot_acf.html
time_lag=20
sm.graphics.tsa.plot_acf(dat,lags=time_lag,ax=ax1)

# Autocorrelation using Kosh fxn
auto_data=np.array([k_autocorrelation(a,k) for k in range(len(a)-1)])
plt.plot(auto_data,linestyle='--',color='k')

# Params
plt.xlim([0,time_lag])
plt.ylim([-1,1])
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
plt.xlabel('Lags',fontsize=5)
plt.ylabel('Correlations',fontsize=5)
plt.title('Autocorrelation for: %s in %s with windows=%s'%(dat_norm_sort.index[i],patient,time_lag),fontsize=5)

# (1) Make a heatmap of thia for each patient.
```

#### Blood de-convolution problem for the microbiome - 

The cell-free microbiome measurments in blood may be compared to matched body site microbiome samples in order to determine the microbial contribution from various tissues to blood. In this case, the vector of bugs measured in blood, $\vec{B}$, is modeled as a linear combination of these same bugs measured in some body site ($j$), $\vec{b_j}$ were $\epsilon$ is the error term:

$$ \vec{B} = h_{\pi} = \sum\limits_j \pi_j \vec{b_j} + \epsilon $$

Our objective is to minimize the squared error (for a single matched blood-body site measurment):

$$ J(\pi) = (\vec{B} - \sum\limits_j \pi_j \vec{b_j})^2 = ( \vec{B} - B_{sites} \cdot \vec{\pi} )^2 $$

Where we solve for $\vec{\pi}$ subject to constraints:

(1) Summation of all tissue fraction leading to the observed contribution is one: $\sum\limits \pi_j = 1 $ to achieve a relative weight to measured body sites.

(2) The contribution from each tissue type: $\pi_j \geq 0 $.

#### Example minimization problem to test the algorithm - 

Note that the R package solves quadratic minimization problems of the form:

$$ min ( -d_{vec}^T b + \frac{1}{2} D_{mat} b^Tb ) = min ( -d_{vec}^T b + \frac{1}{2} D_{mat} b^2 ) $$

Subject to:

$$ A_{mat}^T b \geq b_{vec}$$

Where:

$D_{mat}$ = Matrix in quadratic funtion to be minimize.

$d_{vec}$ = Vector in quadratic funtion to be minimize.

$A_{mat}$ = Matrix defining the constraints under which we want to minimize the quadratic function.

$b_{vec}$ = Initial values of b.

```python
from scipy import stats
import rpy2
import rpy2.robjects as robjects
%load_ext rmagic 

# Objective function 

def obj_fxn(b):
    objective_o=np.dot(-1*dvec,b) + 0.5*np.dot(b.transpose(),b)*Dmat
    return objective_o

dvec=np.array([[0,5,0]]) # Define this as a row vector (actually the transpose of the initial column vector). 
b=np.array([[-8,2,0]])
b_o=b.transpose()
Dmat=np.identity(3)
b_return=obj_fxn(b_o)
b_return
```

#### Re-cast the tissue de-composition problem in a form that can be easily input to quadratic model -

Our objective is to minimize the squared error (for a single matched blood-body site measurment):

$$ min ( \vec{B} - B_{sites} \cdot \vec{\pi} )^2 $$


Re-work the minimization (expand terms) so that it is in a form that allows us to easily parameterize the quadratic model:

$$ -2\vec{B} \cdot B_{sites} \vec{\pi} + B_{sites}^2 \vec{\pi}^2 $$

Subject to equality and inequality constraints:

* $\sum\limits \pi_j = 1 $

* $\pi_j \geq 0 $

---

Recall that R package solves quadratic minimization problems of the form:

$$ min ( -d_{vec}^T b + \frac{1}{2} D_{mat} b^2 ) $$

Subject to:

$$ A_{mat}^T b \geq b_{vec}$$

---

So, we parameterize the model as follows:

$D_{mat}$ = Matrix in quadratic funtion to be minimize = $B_{sites}^T B_{sites}$. 

$d_{vec}$ = Vector in quadratic funtion to be minimize = $\vec{B}^T B_{sites}$.

$A_{mat}$ = Matrix defining the constraints under which we want to minimize the quadratic function. This will have two terms:

Equality constraint: 

* $A_{mat}[0]$ is a vector of one that is the size of $\vec{\pi}$ such that the inner product is computed as expected:
* $(A_{mat}[0])^T \pi = 1$

Inequality constraint: 

* $A_{mat}[1]$ is just the identitify matrix.
* $(A_{mat}[1])^T \vec{\pi} = \vec{\pi} \geq 0$ 

Constraint values:

* $b_{vec}$[0] = 1
* $b_{vec}[1]$ is a vector of one that is the size of $\vec{\pi}$.

#### Indicator (body site specific) bug analysis - 

```python
stor=[]
stor_all_indicators=[]

list_of_samples=data_all_samples_bodysites_matched_tp.index.get_level_values('Sample')
for sample in set(list_of_samples):

    # Sample data
    sam_dat=data_all_samples_bodysites_matched_tp.loc[sample]
    
    # Norm per sample 
    sam_dat_norm_per_sample=sam_dat.div(sam_dat.sum(axis=1),axis=0)
    
    # Perform accounting of the 16s data 
    dat_sample_no_blood=sam_dat_norm_per_sample.drop('Blood')
    dat_sample_no_blood_normPerBug=dat_sample_no_blood.div(dat_sample_no_blood.sum(axis=0),axis=1)
    dat_sample_no_blood_normPerBug.fillna(0,inplace=True)
    dat_sample_no_blood_normPerBug=dat_sample_no_blood_normPerBug[dat_sample_no_blood_normPerBug.sum(axis=1)!=0]
    
    # Threshold for counting tissue specific bugs across 16s sampled body sites
    threshold=0.9
    above_threshold=dat_sample_no_blood_normPerBug[dat_sample_no_blood_normPerBug>threshold]
    above_threshold.fillna(0,inplace=True)
    
    # Mask that will slice out indicator bugs and corresponding measurment of the bug in blood
    ix=above_threshold.sum(axis=0)
    ix=ix[ix>0]
    indexer=above_threshold[ix.index]
    indexer.loc['Blood',:]=1
        
    # Use indexer as a mask to slice data
    sam_dat_norm_per_sample_indicator_bugs=sam_dat_norm_per_sample[indexer>0]
    sam_dat_norm_per_sample_indicator_bugs.fillna(0,inplace=True)
    indicators=sam_dat_norm_per_sample_indicator_bugs[indexer.columns]
    indicators.loc['Sample']=sample
    
    # Store all tissue specific indicators 
    stor_all_indicators=stor_all_indicators+[indicators]
    
    # Store only tissue specific indicator bugs that were also found in blood
    also_in_blood=indicators.loc['Blood']
    to_store=indicators[also_in_blood[also_in_blood>0].index]
    stor=stor+[to_store]
```

```python
def makeBar(df):
    ind=np.arange(df.shape[0])
    plt.bar(ind-0.35,np.array(df),color='k')
    locs,labels=plt.xticks(ind,df.index)
    plt.setp(labels, rotation=90, fontsize=5)
    locs,labels=plt.yticks()
    plt.setp(labels,fontsize=5)
    plt.ylabel('Count',fontsize=5)
    
def quanifyBugs(bs_dat):
    in_bodysite=bs_dat.reset_index()
    specific_bugs=in_bodysite.groupby('tax_name').size()
    specific_bugs.sort(ascending=False)
    return specific_bugs

def quantifySamples(bs_dat):
    in_bodysite=bs_dat.reset_index()
    samples_for_specific_bugs=in_bodysite.groupby('Sample').size()
    samples_for_specific_bugs.sort(ascending=False)
    return samples_for_specific_bugs

# Get all body site specific bugs 
bodysites=list(indicator_bugs_all.index)
bodysites.remove('Sample')
bodysites.remove('Blood')

# Iterate through body sites
for bodysite in bodysites:
    
    print "----"
    print bodysite
    print "----"

    # Get bodysite data
    bs_dat=indicator_bugs_all.loc[bodysite]
    bodysite_specific_bugs=bs_dat[bs_dat>0]
    
     # Check that the body site specific bugs are detected across a range of samples 
    sample_dat=indicator_bugs_all.loc['Sample']
    samples_for_bodysite_specific_bugs=sample_dat[bs_dat>0]
    samples=quantifySamples(samples_for_bodysite_specific_bugs)
    # print samples
    
    # Quantify tissue specific bugs
    df=quanifyBugs(bodysite_specific_bugs)
    plt.subplot2grid((2,2),(0, 0))
    makeBar(df)
    plt.title('All %s specific bugs'%bodysite,fontsize=5)
    
    # Get data for blood for the tissue specific bugs
    blood_dat=indicator_bugs_all.loc['Blood']
    blood_dat_with_buginbs=blood_dat[bs_dat>0]
    in_blood_in_bs=blood_dat_with_buginbs[blood_dat_with_buginbs>0]
    df=quanifyBugs(in_blood_in_bs)
    plt.subplot2grid((2,2),(0,1))
    makeBar(df)
    plt.title('%s also detected in blood (matched sample)'%bodysite,fontsize=5)
    
    # Store all
    store_bug_data=pd.DataFrame()
    store_bug_data[bodysite]=bodysite_specific_bugs
    store_bug_data['Blood']=blood_dat_with_buginbs
    
    # Plot abundance of bug
    plt.subplot2grid((2,2),(1,0),colspan=2)
    uniq_bugs=list(set(store_bug_data.index))
    jet = cm = plt.get_cmap('jet') 
    cNorm = colors.Normalize(vmin=0,vmax=len(uniq_bugs))
    scalarMap = cmx.ScalarMappable(norm=cNorm,cmap=jet) 
    for bug,c in zip(uniq_bugs,range(len(uniq_bugs))):
        to_plot=store_bug_data.loc[bug]
        plt.scatter(to_plot['Blood'],to_plot[bodysite],label=bug,c=scalarMap.to_rgba([c]))
    
    plt.legend(loc='upper right',scatterpoints=1,prop={'size':4})
    plt.axvline(x=0,linestyle='dashed',c='k')
    plt.axhline(y=0,linestyle='dashed',c='k')
    plt.xlim([-.1,1])
    plt.ylim([-.1,1])
    plt.xlabel('Abundance of bug in blood',fontsize=8)
    plt.ylabel('Abundance of \n specific bug in %s'%bodysite,fontsize=5)
    plt.title('Abundance of %s specific bugs relaitve to blood'%bodysite,fontsize=5)
    
    plt.tight_layout()
    plt.show()
```
