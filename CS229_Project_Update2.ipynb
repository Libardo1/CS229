### Read data - 

* Data frame of consolodated 16s and metagenomic data.
* Pruned at genus level.
* With body site and sample_ID as rows.

```python
import sys,os
import pandas as pd

# Read data matched with the blood samples - 
fi = os.getcwd()+'/PR_sample_data_matched_genus.csv'
dat = pd.read_csv(fi, header=[0, 1], index_col=[0, 1])
# Transpose so that bugs are columns and samples are rows and simplify comlumn naming - 
dat_tp=dat.transpose()
names=dat_tp.columns.get_level_values('tax_name')
dat_tp.columns=names
# Swap row levels so that we can easily slice by sample -
dat_tp=dat_tp.swaplevel('Sample', 'BodySite', axis=0)
print "There are %s samples, each of which measures %s bugs."%(dat_tp.shape[0],dat_tp.shape[1])
dat_tp.head(3)
```

```python
# Remove all genus that are ONLY in blood!
dat_lessBlood=dat_tp[dat_tp.index.get_level_values('BodySite')!='Blood']
bug_sums=dat_lessBlood.sum(axis=0)
bug_sums_pruned=bug_sums[bug_sums>0]
# Re-slice the data to remove bugs exclusive to blood
dat_tp=dat_tp[bug_sums_pruned.index]
# Normalized data for each sample 
dat_tp_norm=dat_tp.div(dat_tp.sum(axis=1),axis=0)

print "There are %s samples, each of which measures %s bugs."%(dat_tp_norm.shape[0],dat_tp_norm.shape[1])
dat_tp_norm.head(3)
```

### Supervised Learning: Regression - 

#### Define the problem:

Each blood (or body site) measurment is a vector of bugs (with the same feature space for each measurment):

$$ \vec{B} = \begin{bmatrix} b_1 \\ ... \\ b_{n} \end{bmatrix} $$

$\vec{B}$ is modeled as a linear combination of the measurments across all body sites ($j$) with some error,  $\epsilon$:

$$ \vec{B} = \begin{bmatrix} b_1 \\ ... \\ b_{n} \end{bmatrix} =  \begin{bmatrix} b_{j1} \\ ... \\ b_{jn} \end{bmatrix} \pi_j + . . . + \epsilon $$

$$ \vec{B} = \sum\limits_j \pi_j \vec{b_j} + \epsilon $$

Our objective is to minimize the squared error between model and blood measurment:

$$ J(\pi) = (\vec{B} - \sum\limits_j \pi_j \vec{b_j})^2 = ( \vec{B} - B_{sites} \cdot \vec{\pi} )^2 $$

Subject to constraints:

(1) All tissue fraction leading to the observed contribution is $1$ ($\sum\limits \pi_j = 1$) to obtain a relative weight to measured body sites.

(2) The contribution from each tissue type: $\pi_j \geq 0 $.

---

Our objective is to minimize the squared error (for a single matched blood-body site measurment):

$$ min ( \vec{B} - B_{sites} \cdot \vec{\pi} )^2 $$


Re-work the minimization (expand terms) so that it is in a form that allows us to easily parameterize the quadratic model:

$$ -2\vec{B} \cdot B_{sites} \vec{\pi} + B_{sites}^2 \vec{\pi}^2 $$

Subject to equality and inequality constraints:

* $\sum\limits \pi_j = 1 $

* $\pi_j \geq 0 $

---

Recall that R package solves quadratic minimization problems of the form:

$$ min ( -d_{vec}^T b + \frac{1}{2} D_{mat} b^2 ) $$

Subject to:

$$ A_{mat}^T b \geq b_{vec}$$

So, we parameterize the model as follows:

$D_{mat}$ = Matrix in quadratic funtion to be minimize = $B_{sites}^T B_{sites}$. 

$d_{vec}$ = Vector in quadratic funtion to be minimize = $\vec{B}^T B_{sites}$.

$A_{mat}$ = Matrix defining the constraints under which we want to minimize the quadratic function. This will have two terms:

Equality constraint: 

* $A_{mat}[0]$ is a vector of one that is the size of $\vec{\pi}$ such that the inner product is computed as expected:
* $(A_{mat}[0])^T \pi = 1$

Inequality constraint: 

* $A_{mat}[1]$ is just the identitify matrix.
* $(A_{mat}[1])^T \vec{\pi} = \vec{\pi} \geq 0$ 

Constraint values:

* $b_{vec}$[0] = 1
* $b_{vec}[1]$ is a vector of one that is the size of $\vec{\pi}$.

#### Run regression on the same sample - 

```python
from scipy import stats
import rpy2
import rpy2.robjects as robjects
%load_ext rmagic 
%reload_ext rmagic

def makeBar(df):
    ''' Bar plot function'''
    
    ind=np.arange(df.shape[0])
    plt.bar(ind-0.35,np.array(df),color='k')
    locs,labels=plt.xticks(ind,df.index)
    plt.setp(labels, rotation=90, fontsize=5)
    locs,labels=plt.yticks()
    plt.setp(labels,fontsize=5)
    plt.xlim([-1,df.shape[0]-1])
    plt.ylim([0,1])
    plt.title('Tissue loadings from model',fontsize=8)

def analyzeSample(model_coef,B_sites):
    ''' Model analysis '''
    
    # Store values 
    df=pd.DataFrame()
    df['Per']=np.array(model_coef)
    df['Body sites']=B_sites.columns
    df.set_index('Body sites',inplace=True)
    
    plt.subplot(2,2,1)
    makeBar(df)
    plt.ylabel('Percentage',fontsize=5)
    
    plt.subplot(2,2,2)
    model_guess=(B_sites*model_coef).sum(axis=1)
    plt.scatter(B,model_guess)
    slope, intercept, r_value, p_value, std_err = stats.linregress(B,model_guess)
    plt.xlabel('Blood data',fontsize=5)
    plt.xticks(fontsize=5)
    plt.ylabel('Model guess',fontsize=5)
    plt.yticks(fontsize=5)
    plt.title('Model fit to data with $r^2$=%s'%np.round(r_value**2,decimals=2),fontsize=8)
    
    model_guess.sort()
    B.sort()
        
    plt.tight_layout()
    
    return  (r_value**2,model_guess)

def runQP(test_sample):
    ''' Run QP in R'''
    
    # Model parameters
    B=test_sample.loc['Blood']
    B_sites=test_sample.drop('Blood')
    B_sites=B_sites.transpose()
    tissues=len(B_sites.columns)
    Dmat_in=np.dot(B_sites.transpose(),B_sites)
    dvec_in=np.dot(B.transpose(),B_sites)
    
    # Push parameters into R and run the QP model 
    %Rpush B B_sites tissues dvec_in Dmat_in
    
    # Pull in this package
    %R require(quadprog)
    
    # Body site matrix and blood target
    %R X <- B_sites
    %R Y <- B
    
    # QP constraints and constraint target values 
    %R C <- cbind(rep(1,tissues),diag(tissues))
    %R b <- c(1, rep(0, tissues))
    %R sol <- solve.QP(Dmat = Dmat_in, dvec = dvec_in, Amat = C, bvec = b, meq = 1)
    %Rpull sol
    
    # Model
    model_coef=sol[0]
    
    return (B_sites,model_coef)
```

### Supervised Learning: Classification - 

#### Naive Bayes classifier - 

k-fold CV:
    
* The training set is split into k smaller sets.
* A model is trained using k-1 of the folds as training data.
* The resulting model is validated on the remaining part of the data

Score:

* NB - the mean accuracy on the given test data and labels
