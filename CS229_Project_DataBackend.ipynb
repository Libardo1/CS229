```python
# Build dictionary between body site and sampleID -

sample_to_site_dict=ref_file[['#SampleID','BodySite']]
sample_to_site_dict=sample_to_site_dict.set_index('#SampleID',drop=True)

# Not sampleID in this file can be parsed into its constituent parts - 
ref_file['SubjectID']=ref_file['#SampleID'].apply(lambda x:x[0:5])
ref_file['Spec_type']=ref_file['#SampleID'].apply(lambda x:x[5:7])
ref_file['Gest_week']=ref_file['#SampleID'].apply(lambda x:x[7:9])
ref_file['Bank']=ref_file['#SampleID'].apply(lambda x:x[-1])
ref_file['Sample_ID']='PR'+ref_file['SubjectID']+'_'+ref_file['Gest_week']
ref_file.set_index('#SampleID',inplace=True)

sample_to_site_dict.head(3)
```

```python
# Build dictionary between OTU_ID and taxonomic tree -

def cleanList(l):
    try:
        name=l[1]
    except IndexError:
        name=''
    return name

otu=pd.read_table("/quakestor2/llmartin/infectome/pregnancy/File10b_denovo_otu_table_97ID_TAX_ADDED_MANUALLY.txt",dtype=str)
otu_dict=otu[['OTUId','taxonomy']]
otu_dict=otu_dict.set_index('OTUId',drop=True)
tmp=list(otu_dict['taxonomy'].apply(lambda x: x.split(';')))
taxDat=pd.DataFrame(tmp)
taxDat.index=otu_dict.index
taxDat=taxDat.astype(str)
taxDat=taxDat.apply(lambda x: x.str.split('__').apply(cleanList),axis=1)
taxDat.columns=[['Kingdom','Phylum', 'Class', 'Order', 'Family', 'Genus', 'Species']]
taxDat.head(3)
```

```python
# Pull NCBI tax table - 

FullTaxTable = pd.read_table("/quakestor2/kertesz/soup/Data/NCBI/tax/Taxtastic/min.tax.table",dtype=str)
FullTaxTable.set_index('tax_id',inplace=True,drop=False)
FullTaxTable.head(3)
```

```python
# Merge OTU and taxTable - 

# Choose the same taxonomic level for OTU and NCBI data - 
taxLevel_otuTable='Genus'
tmp_OTUs=pd.DataFrame(taxDat[taxLevel_otuTable])
tmp_OTUs.reset_index(drop=False,inplace=True)

taxLevel_taxTable='genus'
tmp_ncbiTree=FullTaxTable[FullTaxTable['rank']==taxLevel_taxTable]
tmp_ncbiTree=tmp_ncbiTree[['tax_name','tax_id']]

# Merge these based upon tax name, giving us a link between OTU_ID and tax_id, which we need later - 
taxID_to_OTU_mapping=pd.merge(tmp_OTUs,tmp_ncbiTree,left_on=taxLevel_otuTable,right_on="tax_name")
taxID_to_OTU_mapping.head(3)
```

```python
# Pull and annotate taxID data - 

# Pull OTU data table, merge with table above, which extracts all GENUS (specified above) and provides associated tax_ID [columns = sample IDs] - 
otu=pd.read_table("/quakestor2/llmartin/infectome/pregnancy/File10b_denovo_otu_table_97ID_TAX_ADDED_MANUALLY.txt",dtype=str)
otu=otu.set_index('OTUId',drop=False) 
samples_wOTU_andTax=pd.merge(otu,taxID_to_OTU_mapping,left_on='OTUId',right_on="OTUId")

# Add body site data for each sample as a row [body site for each OTU sample_ID] - 
def getBodySite(sampleID):
    try:
        bs=sample_to_site_dict.loc[sampleID][0]
    except KeyError:
        bs=''
    return bs
b_sites=[getBodySite(col) for col in samples_wOTU_andTax.columns]
samples_wOTU_andTax.loc['bodysite']=b_sites

# Add "universal" sampleID (the one we use in cell-free data) to each OTU sample_ID as a row - 
def getSamID(col):
    try:
        sampleID=ref_file.loc[col,'Sample_ID']
    except KeyError:
        sampleID=col
    return sampleID
sam_names=[getSamID(col) for col in samples_wOTU_andTax.columns]
samples_wOTU_andTax.loc['Sample_ID']=sam_names

# Transpose so that OTUs are columns and samples are rows - 
samples_wOTU_andTax_tp=samples_wOTU_andTax.transpose()

# Extract taxIDs assocaited with each OTUid - 
taxID_list=samples_wOTU_andTax_tp.loc['tax_id']

# Replace OTU sample_ID with bodysite and universal sample_ID as the row index - 
samples_wOTU_andTax_tp.set_index(['Sample_ID','bodysite'],drop=False,inplace=True)

# Replace OTUs with taxIDs as the column index - 
samples_wOTU_andTax_tp.columns=taxID_list

# Now we have a dataframe of [sample_ID x tax_ID]
samples_wOTU_andTax_tp.head(3)
```

```python
def get_sample_data(patient_to_analyze):

    # --- Get cell-free data for patient --- 

    # The OTU data is pruned at a specified taxonomic level (specified above).
    # Make sure cell-free data is also pruned at this level.
    # This is because we need to merge on taxIDs --- so this must be consistent.
    level=taxLevel_taxTable 
    path="/quakestor2/kertesz/soup/Data/NCBI/tax/Taxtastic/min.tax.table"
    FullTaxTable=pd.read_table(path,dtype=str)[["tax_id","tax_name",level]]
    FullTaxTable.set_index('tax_id',inplace=True,drop=False)
    mm=pd.merge(GRAMMyAnnot,FullTaxTable,left_on="TaxID",right_on="tax_id")
    mm1=mm[['Normalized_Abundance','Sample_ID',level]]
    counts=mm1.groupby(['Sample_ID',level],as_index=False).aggregate(np.sum)
    counts.set_index('Sample_ID',inplace=True)
    
    # Slice cell-free data for the patient of interest.
    blood_data_toConcat=counts.loc[patient_to_analyze]
    scaleFactor=1000000
    blood_data_toConcat['Normalized_Abundance']=blood_data_toConcat['Normalized_Abundance']*scaleFactor
    blood_data_toConcat.set_index(level,inplace=True)
    blood_data_toConcat.columns=['Blood']
    
    # --- Get 16s data for patient --- 
    
    # 16s data for patient. 
    try:
        patient_16s_data=samples_wOTU_andTax_tp.loc[patient_to_analyze]
    except KeyError:
        #print "Correct 16s data for %s"%patient_to_analyze
        patient_16s_data=samples_wOTU_andTax_tp.loc[correction_dict[patient_to_analyze]]
            
    # List for storage.
    storeData=[]
    
    # Iterate through body sites for the given patient.  
    for bodysite in list(set(patient_16s_data.index)):
        
        # This returns a series indexed by taxID, or a dataframe (if > 1 sample).
        patient_16s_data_bodysite=patient_16s_data.loc[bodysite]
        
        try:
            # Sometimes samples will have more than one body site entiry. 
            # Presumbly two samples taken and sequenced. 
            # In this case, the Dataframe will have two dimentions (as opposed to a series). 
            # Iterate through the individual entires and process.
            flag=patient_16s_data_bodysite.shape[1]
            #print "Duplicate entries for %s and bodysite %s"%(patient_to_analyze,bodysite)
            
            # This stores data from the entries. 
            tmp_stor=[]
            for i in range(patient_16s_data_bodysite.shape[0]):
                # This returns a series indexed by taxID.
                patient_16s_data_bodysite_dupEntry=patient_16s_data_bodysite.ix[i]
                patient_16s_data_bodysite_dupEntry.drop('',inplace=True)
                patient_16s_data_bodysite_dupEntry.drop('tax_id',inplace=True)
                patient_16s_data_bodysite_dupEntry=patient_16s_data_bodysite_dupEntry.astype(float)
                # Aggregate data for multiple instances of a given (genus level) taxID. 
                patient_16s_data_bodysite_grouped=patient_16s_data_bodysite_dupEntry.groupby(patient_16s_data_bodysite_dupEntry.index).sum()
                patient_16s_toConcat=pd.DataFrame(patient_16s_data_bodysite_grouped)
                tmp_stor=tmp_stor+[patient_16s_toConcat]
            tmp_concat=pd.concat(tmp_stor,axis=1)
            
            # Take the mean from duplicate body site measurments - 
            tmp_concat_toConcat=pd.DataFrame(tmp_concat.mean(axis=1))
            tmp_concat_toConcat.columns=[bodysite]
            #print "Shape of pruned %s level 16s data: %s"%(level,tmp_concat_toConcat.shape[0])
            
            storeData=storeData+[tmp_concat_toConcat]
        
        # If there is a single body site measurment, Indexerror above is thrown. Process the series in isolation.
        except IndexError:
            patient_16s_data_bodysite.drop('',inplace=True)
            patient_16s_data_bodysite.drop('tax_id',inplace=True)
            patient_16s_data_bodysite=patient_16s_data_bodysite.astype(float)
            patient_16s_data_bodysite_grouped=patient_16s_data_bodysite.groupby(patient_16s_data_bodysite.index).sum()
            patient_16s_toConcat=pd.DataFrame(patient_16s_data_bodysite_grouped)
            patient_16s_toConcat.columns=[bodysite]
            #print "Shape of pruned %s level 16s data: %s"%(level,patient_16s_toConcat.shape[0])
            storeData=storeData+[patient_16s_toConcat]
    
    # List of data from all body sites for samples. Add the corresponding cell-free DNA (blood) data. 
    storeData=storeData+[blood_data_toConcat]
    data_all_cat=pd.concat(storeData,axis=1)
    data_all_cat.fillna(0,inplace=True)
    data_all_cat_norm=data_all_cat
    
    # Sort data by load 
    sums=data_all_cat_norm.sum(axis=1)
    sums.sort(ascending=False)
    data_all_cat_norm=data_all_cat_norm.loc[sums.index]
    #print "Shape of pruned %s level condolodated blood and 16s data: %s"%(level,data_all_cat_norm.shape[0])
        
    # Specify final level of aggregation desired.
    # Class is nice beacuse it make viewing the data easier (fewer taxonomic categories) later.
    level_to_view_data='genus'
    FullTaxTable=pd.read_table(path,dtype=str)[["tax_id","tax_name",level_to_view_data]]
    FullTaxTable.set_index('tax_id',inplace=True,drop=False)
    merge_data=pd.merge(data_all_cat_norm,FullTaxTable,left_on=data_all_cat_norm.index,right_on="tax_id")
    merge_data.set_index(['tax_id','tax_name'],inplace=True)
    data_abs=merge_data.groupby([level_to_view_data],as_index=False).aggregate(np.sum)
    data_abs.set_index(level_to_view_data,inplace=True)
    data_abs_annot=pd.merge(data_abs,FullTaxTable,left_on=data_abs.index,right_on="tax_id")
    #print "Shape of final pruned %s level condolodated blood and 16s data: %s"%(level,data_abs_annot.shape[0])
        
    return data_abs_annot,level_to_view_data
```

#### Process matched samples - 

```python
# This will store cfDNA data with matched 16s data --- 
tmp_store=[]
for sample in list(set(GRAMMyAnnot['Sample_ID'])):
    
    # Cleanup for concatenation
    dat_sample,level_used_for_final_prune=get_sample_data(sample)
    try:
        
        # Prepare sample data
        dat_sample.set_index(['tax_id','tax_name'],inplace=True)
        del dat_sample[level_used_for_final_prune]
        dat_sample_norm=dat_sample.div(dat_sample.sum(axis=0),axis=1)
        dat_sample_norm.fillna(0,inplace=True)
        
        # Add sample name to the index 
        dat_sample_tp=dat_sample.transpose()
        dat_sample_tp['Sample']=sample
        dat_sample_tp.set_index('Sample',append=True,inplace=True,drop=True)
        dat_sample_withName=dat_sample_tp.transpose()
        
        # Store
        tmp_store=tmp_store+[dat_sample_withName]
        
    except AttributeError:
        None
        # print "No 16s data for %s"%sample
        
# Concatenate data - 
data_all_samples_bodysites=pd.concat(tmp_store,axis=1)
data_all_samples_bodysites.fillna(0,inplace=True)
```

```python
# --- This will store all 16s data from the patients in which cfDNA is also sampled --- 

samples=pd.Series(samples_wOTU_andTax_tp.index.get_level_values('Sample_ID'))
sample_IDs=samples[samples.str.contains('PR')]

# --- Unique sample names to iterate through ---

# Get all 16s samples for patients in which cfDNA was also sampled --- 
patients_with_cfData=list(set([sam.split('_')[0] for sam in list(set(GRAMMyAnnot['Sample_ID']))]))
sample_list=list(samples_wOTU_andTax_tp.index.get_level_values('Sample_ID'))
relevant_16s_samples=[sam for sam in sample_list if sam.split('_')[0] in patients_with_cfData]

# Combine this with the timepoint for cfDNA samples --- 
list_of_cfDNA_samples=list(set(GRAMMyAnnot['Sample_ID']))
consolodated_samples_for_patients=list(set(list_of_cfDNA_samples+relevant_16s_samples))

# The OTU data is pruned at a specified taxonomic level (specified above). Group the cfDNA data at the same level.
level=taxLevel_taxTable 
path="/quakestor2/kertesz/soup/Data/NCBI/tax/Taxtastic/min.tax.table"
FullTaxTable=pd.read_table(path,dtype=str)[["tax_id","tax_name",level]]
FullTaxTable.set_index('tax_id',inplace=True,drop=False)
mm=pd.merge(GRAMMyAnnot,FullTaxTable,left_on="TaxID",right_on="tax_id")
mm1=mm[['Normalized_Abundance','Sample_ID',level]]
counts_cfDNA=mm1.groupby(['Sample_ID',level],as_index=False).aggregate(np.sum)
counts_cfDNA.set_index('Sample_ID',inplace=True)

# Merge data
def merge_16s_with_cfDNA(sample_to_analyze):

     # Get cfDNA for sample - 
    flag_cf=1
    try:
        blood_data_toConcat=counts_cfDNA.loc[sample_to_analyze]
        scaleFactor=1000000
        blood_data_toConcat['Normalized_Abundance']=blood_data_toConcat['Normalized_Abundance']*scaleFactor
        blood_data_toConcat.set_index(level,inplace=True)
        blood_data_toConcat.columns=['Blood']
    except:
        flag_cf=0
        #print "No cfDNA data for patient: %s"%sample_to_analyze
        blood_data_toConcat=[]
        
    # Get 16s data for sample - 
    flag_16s=1
    try:
        patient_16s_data=samples_wOTU_andTax_tp.loc[sample_to_analyze]
    except KeyError:
        #print "Correct 16s data for %s"%sample_to_analyze
        patient_16s_data=[]
        flag_16s=0

    # List for storage of 16s data - 
    storeData=[]
        
    # If there is 16s data, then iterate through body sites - 
    if flag_16s:
    
        # Iterate through body sites for the given patient.  
        for bodysite in list(set(patient_16s_data.index)):

            # This returns a series indexed by taxID, or a dataframe (if > 1 sample).
            patient_16s_data_bodysite=patient_16s_data.loc[bodysite]

            try:
                # Sometimes samples will have more than one body site entiry. 
                # Presumbly two samples taken and sequenced. 
                # In this case, the Dataframe will have two dimentions (as opposed to a series). 
                # Iterate through the individual entires and process.
                flag=patient_16s_data_bodysite.shape[1]
                # print "Duplicate entries for %s and bodysite %s"%(patient_to_analyze,bodysite)

                # This stores data from the entries. 
                tmp_stor=[]
                for i in range(patient_16s_data_bodysite.shape[0]):
                    # This returns a series indexed by taxID.
                    patient_16s_data_bodysite_dupEntry=patient_16s_data_bodysite.ix[i]
                    patient_16s_data_bodysite_dupEntry.drop('',inplace=True)
                    patient_16s_data_bodysite_dupEntry.drop('tax_id',inplace=True)
                    patient_16s_data_bodysite_dupEntry=patient_16s_data_bodysite_dupEntry.astype(float)
                    # Aggregate data for multiple instances of a given taxID.
                    patient_16s_data_bodysite_grouped=patient_16s_data_bodysite_dupEntry.groupby(patient_16s_data_bodysite_dupEntry.index).sum()
                    patient_16s_toConcat=pd.DataFrame(patient_16s_data_bodysite_grouped)
                    tmp_stor=tmp_stor+[patient_16s_toConcat]
                tmp_concat=pd.concat(tmp_stor,axis=1)

                # Take the mean from duplicate body site measurments - 
                tmp_concat_toConcat=pd.DataFrame(tmp_concat.mean(axis=1))
                tmp_concat_toConcat.columns=[bodysite]
                storeData=storeData+[tmp_concat_toConcat]

            # If there is a single body site measurment, Indexerror above is thrown. Process the series in isolation.
            except IndexError:
                patient_16s_data_bodysite.drop('',inplace=True)
                patient_16s_data_bodysite.drop('tax_id',inplace=True)
                patient_16s_data_bodysite=patient_16s_data_bodysite.astype(float)
                patient_16s_data_bodysite_grouped=patient_16s_data_bodysite.groupby(patient_16s_data_bodysite.index).sum()
                patient_16s_toConcat=pd.DataFrame(patient_16s_data_bodysite_grouped)
                patient_16s_toConcat.columns=[bodysite]
                storeData=storeData+[patient_16s_toConcat]
    
    if flag_16s and flag_cf:
        # List of data from all body sites for samples. Add the corresponding cell-free DNA (blood) data. 
        storeData=storeData+[blood_data_toConcat]
        data_all_cat=pd.concat(storeData,axis=1)
        data_all_cat.fillna(0,inplace=True)
        data_all_cat_norm=data_all_cat
    
    elif flag_16s:
        data_all_cat=pd.concat(storeData,axis=1)
        data_all_cat.fillna(0,inplace=True)
        data_all_cat_norm=data_all_cat
        
    else:
        data_all_cat=blood_data_toConcat
        data_all_cat.fillna(0,inplace=True)
        data_all_cat_norm=data_all_cat
        
    # Sort data by load 
    sums=data_all_cat_norm.sum(axis=1)
    sums.sort(ascending=False)
    data_all_cat_norm=data_all_cat_norm.loc[sums.index]
        
    # Specify final level of aggregation desired (class is nice beacuse it make viewing the data easier).
    level_to_view_data='genus'
    FullTaxTable=pd.read_table(path,dtype=str)[["tax_id","tax_name",level_to_view_data]]
    FullTaxTable.set_index('tax_id',inplace=True,drop=False)
    merge_data=pd.merge(data_all_cat_norm,FullTaxTable,left_on=data_all_cat_norm.index,right_on="tax_id")
    merge_data.set_index(['tax_id','tax_name'],inplace=True)
    data_abs=merge_data.groupby([level_to_view_data],as_index=False).aggregate(np.sum)
    data_abs.set_index(level_to_view_data,inplace=True)
    data_abs_annot=pd.merge(data_abs,FullTaxTable,left_on=data_abs.index,right_on="tax_id")
        
    return data_abs_annot,level_to_view_data

# This will store all cfDNA data with matched 16s data --- 
tmp_store=[]

for sample in consolodated_samples_for_patients:
    
    # Cleanup for concatenation
    dat_sample,level_to_view_data=merge_16s_with_cfDNA(sample)
    try:
        
        # Prepare sample data
        dat_sample.set_index(['tax_id','tax_name'],inplace=True)
        del dat_sample[level_to_view_data]
        dat_sample_norm=dat_sample.div(dat_sample.sum(axis=0),axis=1)
        dat_sample_norm.fillna(0,inplace=True)
        
        # Add sample name to the index 
        dat_sample_tp=dat_sample.transpose()
        dat_sample_tp['Sample']=sample
        dat_sample_tp.set_index('Sample',append=True,inplace=True,drop=True)
        dat_sample_withName=dat_sample_tp.transpose()
        
        # Store
        tmp_store=tmp_store+[dat_sample_withName]
        
    except AttributeError:
        None 
        # print "No 16s data for %s"%sample
        
# Concatenate data - 
data_all_16s_samples_bodysites=pd.concat(tmp_store,axis=1)
data_all_16s_samples_bodysites.fillna(0,inplace=True)
```

```python
def makeSampleDict(pathToDict):
    d={}
    with open(pathToDict, 'r') as infile:
        for line in infile:
            # First field (comma delimited) is the name	
            sampleName=line.rstrip().split(',')[0].strip('"')
            # Second field (comma delimited) is the body site	
            bodySite=line.rstrip().split(',')[1].strip('"')
            d[sampleName]=bodySite
    return d

def getBodySites(path):
    # Dictionary mapping HMP sample name to body site
    pathToDict=path+"/sample_bodySite"
    sampleToBodySite=makeSampleDict(pathToDict)
    # Dictionary mapping HMP sample name to body site
    sampleNames=path+"/HMP_sample_names"
    samplesNames=np.genfromtxt(sampleNames,dtype=str,delimiter='\t')
    bodySites=[]
    for name in samplesNames:
        try:
            site=sampleToBodySite[name]
            bodySites=bodySites+[site]
        except KeyError:
            flag=0
    return  bodySites

# Column body site labels.
path=os.getcwd()+'/reference/'
HMPdataBodySites=list(getBodySites(path))

# Column of microbes and matrix of abundances.
hmpData=path+"HMP_rawdata"
hmpDF=pd.DataFrame(pd.read_table(hmpData,header=None))
hmpDF=hmpDF.set_index([0]) # Rows are bugs (718 total)
hmpDF.columns=HMPdataBodySites # Columns are body sites for each sample (690)
hmpDF.index.name='Bug'
hmpDF.head(3)
```
