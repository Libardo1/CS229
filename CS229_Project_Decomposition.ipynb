### Read data - 

Approach: 

* Pull data frame of consolodated 16s and shotgun sequencing data pruned at genus level.
* Select bugs measured in the sampled 16s body sites and the blood, as blood (shotgun sequencing) measurments will not exclusivly measure microbes.
* Normalize such that each sample is a vector of bug abundances that sum to 1.

```python
import sys,os
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt 
%matplotlib inline
matplotlib.rcParams['savefig.dpi'] = 3 * matplotlib.rcParams['savefig.dpi']

# Read data matched with the blood samples - 
fi = os.getcwd()+'/PR_sample_data_matched_genus.csv'
dat = pd.read_csv(fi, header=[0, 1], index_col=[0, 1])

# Transpose so that bugs are columns and samples are rows and simplify comlumn naming - 
dat_tp=dat.transpose()
names=dat_tp.columns.get_level_values('tax_name')
dat_tp.columns=names

# Swap row levels so that we can easily slice by sample -
dat_tp=dat_tp.swaplevel('Sample', 'BodySite', axis=0)

# Remove all genus that are exclusive to blood, because metagenomic sequencing will be richer in this respect than 16s - 
dat_lessBlood=dat_tp[dat_tp.index.get_level_values('BodySite')!='Blood']
bug_sums=dat_lessBlood.sum(axis=0)
bug_sums_pruned=bug_sums[bug_sums>0]

# Re-slice the data to remove bugs exclusive to blood -
dat_tp=dat_tp[bug_sums_pruned.index]

# Normalized data for each sample - 
dat_tp_norm=dat_tp.div(dat_tp.sum(axis=1),axis=0)

print "There are %s samples with genus shared between blood / body sites, each of which measures %s bugs."%(dat_tp_norm.shape[0],dat_tp_norm.shape[1])
```

### Compute mean abundances - 

First, compute the mean abundance of each bug at each body site:
    
* Slice the data by body site, which has $n$ samples. This results in $n \times k$ matrix. 
* Compute mean abunance across $n$ samples for each of the $k$ bugs.
* This results a $1 \times k$ vector of mean abundances.
* Perform this for each of $m$ body sites, resulting in:

$$ b_{means} = \begin{bmatrix} b_{11} ... b_{1m} \\ ... \\  b_{k1}  ... b_{km} \end{bmatrix} $$

* Thus, we get an array of mean bug abundance with respect to body sites.

```python
# Slice by body site
stor=[]
for bodysite in set(dat_tp.index.get_level_values('BodySite')):
    # Slice data, givin n measurments x k bugs
    site_dat=dat_tp_norm[dat_tp_norm.index.get_level_values('BodySite')==bodysite]
    # Mean for each per bug
    site_mean=site_dat.mean(axis=0)
    # Store
    frac_bug_toagg=pd.DataFrame(site_mean,columns=[bodysite])
    stor=stor+[frac_bug_toagg]
bs_dat=pd.concat(stor,axis=1)
bs_dat.fillna(0,inplace=True)

# Condense the oral data
sums=bs_dat.sum(axis=1)
sums.sort(ascending=False)
bs_dat=bs_dat.loc[sums.index]
bs_dat['Oral']=(bs_dat['Saliva']+bs_dat['Tooth_Gum'])/2

# Delete columns
del bs_dat['CordBlood_Serum']
del bs_dat['Saliva']
del bs_dat['Tooth_Gum']
```

### Linear regression to evaluate body site contribution - 

Let's take this one step further to try and further de-convolue the the two possible models:

* We can do this by examining abundances for each sample individually, rather than examining mean abundances (above).
* For each sample (e.g., a timepoint), we have a vector of bug abundances for (1) various tissues and (2) blood.
* In this case, "blind sources" contribute bugs found in blood, but not found in any matched body sites at that timepoint.

Let's assume a linear model:

* The load of bugs in blood is a linear combination of bugs measured at each body site.
* Include a "blind source," which includes bugs that are present in blood but not measured in any tissues for the sample.
* Solve for the model parameters (e.g., tissue weights) that minimize the squared error between model and blood measurment.
* Note that this is subject to some reasonable constraints (detailed below) that nessecate use of a QP solver.

#### Define the problem:

Each blood (or body site) measurment is a vector of bugs (with the same feature space for each measurment):

$$ \vec{B} = \begin{bmatrix} b_1 \\ ... \\ b_{n} \end{bmatrix} $$

$\vec{B}$ is modeled as a linear combination of the measurments across all body sites ($j$) with some error,  $\epsilon$:

* Let $\epsilon$ represent the blind source.
* This is $\epsilon$ are simply the load of bugs observed in blood that are not present in any of the sampled tissued.

$$ \vec{B} = \begin{bmatrix} b_1 \\ ... \\ b_{n} \end{bmatrix} =  \begin{bmatrix} b_{j1} \\ ... \\ b_{jn} \end{bmatrix} \pi_j + . . . + \pi_{\epsilon} \epsilon $$

$$ \vec{B} = \sum\limits_j \pi_j \vec{b_j} + \pi_{\epsilon} \epsilon $$

Our objective is to minimize the squared error between model and blood measurment:

$$ J(\pi) = (\vec{B} - \sum\limits_j \pi_j \vec{b_j})^2 = ( \vec{B} - B_{sites} \cdot \vec{\pi} )^2 $$

Subject to constraints:

(1) All tissue fraction leading to the observed contribution is $1$ ($\sum\limits \pi_j = 1$) to obtain a relative weight to measured body sites.

(2) The contribution from each tissue type: $\pi_j \geq 0 $.

---

Our objective is to minimize the squared error (for a single matched blood-body site measurment):

$$ min ( \vec{B} - B_{sites} \cdot \vec{\pi} )^2 $$


Re-work the minimization (expand terms) so that it is in a form that allows us to easily parameterize the quadratic model:

$$ -2\vec{B} \cdot B_{sites} \vec{\pi} + B_{sites}^2 \vec{\pi}^2 $$

Subject to equality and inequality constraints:

* $\sum\limits \pi_j = 1 $

* $\pi_j \geq 0 $

---

Recall that R package solves quadratic minimization problems of the form:

$$ min ( -d_{vec}^T b + \frac{1}{2} D_{mat} b^2 ) $$

Subject to:

$$ A_{mat}^T b \geq b_{vec}$$

So, we parameterize the model as follows:

$D_{mat}$ = Matrix in quadratic funtion to be minimize = $B_{sites}^T B_{sites}$. 

$d_{vec}$ = Vector in quadratic funtion to be minimize = $\vec{B}^T B_{sites}$.

$A_{mat}$ = Matrix defining the constraints under which we want to minimize the quadratic function. This will have two terms:

Equality constraint: 

* $A_{mat}[0]$ is a vector of one that is the size of $\vec{\pi}$ such that the inner product is computed as expected:
* $(A_{mat}[0])^T \pi = 1$

Inequality constraint: 

* $A_{mat}[1]$ is just the identitify matrix.
* $(A_{mat}[1])^T \vec{\pi} = \vec{\pi} \geq 0$ 

Constraint values:

* $b_{vec}$[0] = 1
* $b_{vec}[1]$ is a vector of one that is the size of $\vec{\pi}$.

Recall our model:

* $\epsilon$ is a vector of bug abundances that are observed in blood, but not present in any of the tissues.

$$ \vec{B} = \begin{bmatrix} b_1 \\ ... \\ b_{n} \end{bmatrix} =  \begin{bmatrix} b_{j1} \\ ... \\ b_{jn} \end{bmatrix} \pi_j + . . . + \pi_{\epsilon} \epsilon $$

* If $\pi_{\epsilon}$ is weighted heavily, it indicates that this "blind source" contributes strongly to the model.
* The fitted weights for each sample are plotted in the heatmap (left) and associated model $r^2$ value on the right.
* In most cases, the blind source ($\pi_{\epsilon}$) is weighted heavily in the optimization.

Therefore:

* In the "indicator" analysis (using mean abundances), we obseved that blood contains bugs that are low abundance in sampled tissues.
* We couldn't de-convolute the (1) non-linear model from the (2) "blind source" model.
* In this analysis, we fit data for each sample individually and therefore clearly define the "blind source" bugs in each case.
* We see that (1) "blind source" are often weighed heavily.
* This argues against the (2) non-linear model and points to the fact that blood is sampling more regions than we sampled. 

### ICA for blind source seperation - 

Approach:

* Can we use a method to estimate these blind sources?
* One approach is ICA.
* We should be able to correlate the recovered signal with our body site measurments.
* Further, we will recover signals of "unknown" (blind) source, which we may be able to correlate with other microbiome literature.

ICA sources:

* A cell is governed by $M$ processes.
* Blood is a mixture of $M$ tissues.

Measurment dimentionality:

* Each process has an assocaited expression level of $K$ genes.
* Each tissue has an assocaited microbiome of $K$ bugs.
* For example, each source that feeds into blood is a vector of $K$ bugs:

$ \vec{s_1} = \begin{bmatrix} b_1 \\ ... \\ b_k \end{bmatrix} $

* Moreover, we have a matrix of source data with dimentionality $M \times K $:

$ S = \begin{bmatrix} s_1 \\ ... \\ s_M \end{bmatrix} $

ICA mixtures:

* Measurments via microarray were made across $N$ different conditions.
* Measurments via cfDNA-sequencing were made across $N$ different conditions.
* In either case, we have an array of measurments:

$ X = \begin{bmatrix} x_1 \\ ... \\ x_N \end{bmatrix} $

* The measured load of gene expression (or bugs) at each condition, $x_j$, is a linear combination of the $M$ sources:

$ x_j = a_{j1} s_1 + ... + a_{jM} s_M $

The model in matrix notation:

$$ \begin{bmatrix} x_1 \\ ... \\ x_N \end{bmatrix} =  \begin{bmatrix} a_{11} ... a_{1M} \\ ... \\  a_{N1}  ... a_{NM} \end{bmatrix} \begin{bmatrix} s_1 \\ ... \\ s_M \end{bmatrix} $$

Assumption:

* The underlying $M$ sources are independent. 

Goal:

* Find a matrix W that provides the transformation $Y=WX$ of the observed data.
* $ Y = WX $, where $Y$ are the independent components.
* The components $(y_1 ... y_M)$ are close approximations of the sources, $(s_1 ... s_M)$

Input:

* Matrix X, which is $N \times K$.
* Re-express this as a mixture of $M$ independent components, $\hat{y}$
* Each element in $\hat{y}$ is a mixture of $K$ loads.


```python
from pylab import *
from numpy import *
from sklearn.decomposition import FastICA

def runICA(dat_tp_norm,ica_comp):
    ''' Run ICA with specified number of components''' 
    
    # Mean for body sites in DF sampled
    mean_dat=pd.DataFrame()
    X_nonblood=dat_tp_norm[dat_tp_norm.index.get_level_values('BodySite')!='Blood']
    for bs in set(X_nonblood.index.get_level_values('BodySite')):
        bs_dat=dat_tp_norm[dat_tp_norm.index.get_level_values('BodySite')==bs]
        means=bs_dat.mean(axis=0)
        means_norm=means/means.sum()
        mean_dat[bs]=means_norm
     
    # Blood data in DF sampled
    X=dat_tp_norm[dat_tp_norm.index.get_level_values('BodySite')=='Blood']
    
    # ICA
    ica = FastICA(n_components=ica_comp)
    ica.fit(X.T)
    
    # Write computed source data to DF
    S_est_bugs = ica.transform(X.T)
    ica_sources=pd.DataFrame()
    for i, si in enumerate(S_est_bugs.T):
        ica_sources['Source_%s'%str(i+1)]=si
    
    return (ica_sources,mean_dat)

def computeCorrl(ica_sources,mean_dat,plotDat):
    ''' Compute correlations between ICA components and mean body site data for same samples.'''
    
    ica_source_coorl=ica_sources.transpose()
    mean_dat_coorl=mean_dat.transpose()
    stor=[] 
    for source_num,i in zip(ica_source_coorl.index,np.arange(len(ica_source_coorl.index))):
        # ICA data
        ica_source=ica_source_coorl.loc[source_num]
        # Store r^2 for the source
        r2_header=source_num+'_R2'
        tmp_stor_r2=pd.DataFrame(columns=[r2_header])
        # Compute r^2 between input source and body sites
        for bs in set(mean_dat_coorl.index):
            bs_dat=mean_dat_coorl.loc[bs]
            slope, intercept, r_value, p_value, std_err = stats.linregress(bs_dat,ica_source)
            tmp_stor_r2.loc[bs]=r_value**2
        stor=stor+[tmp_stor_r2]
        # Plot
        if plotDat: # Plot data
            plt.subplot(3,1,1)
            plot(ica_source, 'k', alpha=0.5, label='ICA source', drawstyle='steps') 
            # Body site with highest corrl 
            tmp_stor_r2_sort=tmp_stor_r2.sort(r2_header,ascending=True)
            bs_top=tmp_stor_r2_sort.index[-1] 
            plot(mean_dat_coorl.loc[bs_top], 'r', alpha=0.5, label='Body site: %s'%bs_top, drawstyle='steps') 
            plt.title('%s with best $r^2=$ %s from %s'%(source_num,np.round(tmp_stor_r2_sort.loc[bs_top].astype(float),decimals=2)[0],bs_top),fontsize=7.5)
            plt.xlim([0,mean_dat_coorl.loc[bs_top].shape[0]])
            plt.ylim([-1.5,1.5])
            plt.legend(fontsize=5)
            plt.xticks(fontsize=5)
            plt.yticks(fontsize=5)
            
            ica_source.sort(ascending=False)
            plt.subplot(3,1,2)
            ind=np.arange(ica_source.shape[0])+0.5
            plt.bar(ind,ica_source)
            locs,labels=plt.xticks(ind,ica_source.index)
            plt.setp(labels, rotation=90, fontsize=3.5)
            plt.xlim([0,len(ica_source.index)]) 
            plt.xticks(fontsize=5)
            plt.yticks(fontsize=5)
            
            plt.tight_layout()
            plt.show()
        
    # Concat all coorleation data
    all_r2=pd.concat(stor,axis=1)
    all_r2=all_r2.astype('float')
    all_r2=np.round(all_r2,decimals=4)
    return all_r2

# Compute ICA
ica_sources,mean_dat=runICA(dat_tp_norm,10)
# Evaluate correlations
all_r2=computeCorrl(ica_sources,mean_dat,0)
sums=all_r2.sum(axis=0)
sums.sort(ascending=False)
all_r2=all_r2[sums.index]
# Plot heatmap
plt.subplot(2,2,1)
makeHeatmap(all_r2)
plt.colorbar()
```
