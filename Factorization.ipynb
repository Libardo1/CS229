```python
import sys,os
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt 
%matplotlib inline
matplotlib.rcParams['savefig.dpi'] = 3 * matplotlib.rcParams['savefig.dpi']

# Read data matched with the blood samples - 
fi = os.getcwd()+'/PR_sample_data_matched_genus.csv'
dat = pd.read_csv(fi, header=[0, 1], index_col=[0, 1])

# Transpose so that bugs are columns and samples are rows and simplify comlumn naming - 
dat_tp=dat.transpose()
names=dat_tp.columns.get_level_values('tax_name')
dat_tp.columns=names

# Swap row levels so that we can easily slice by sample -
dat_tp=dat_tp.swaplevel('Sample', 'BodySite', axis=0)

# Remove all genus that are exclusive to blood, because metagenomic sequencing will be richer in this respect than 16s - 
dat_lessBlood=dat_tp[dat_tp.index.get_level_values('BodySite')!='Blood']
bug_sums=dat_lessBlood.sum(axis=0)
bug_sums_pruned=bug_sums[bug_sums>0]

# Re-slice the data to remove bugs exclusive to blood -
dat_tp=dat_tp[bug_sums_pruned.index]

# Normalized data for each sample - 
dat_tp_norm=dat_tp.div(dat_tp.sum(axis=1),axis=0)

print "There are %s samples with genus shared between blood / body sites, each of which measures %s bugs."%(dat_tp_norm.shape[0],dat_tp_norm.shape[1])
```

### Non-negative matrix factorization - 

http://scikit-learn.org/stable/modules/decomposition.html

$$ V = WH$$

Where:
    
* V is $n \times n $
* W is $m \times p $
* H is $p \times n $

```python
def ICApreprocess(df):
    ''' Zero out the mean, and made stv one. '''
    # Normalize data by samples with zero mean 
    df_mean_norm=df.subtract(df.mean(axis=1),axis=0)
    # Normalize data by samples with unit variance 
    df_mean_var_norm=df_mean_norm.divide(df_mean_norm.std(axis=1),axis=0)
    return df_mean_var_norm

# Blood data in DF sampled
X=dat_tp_norm[dat_tp_norm.index.get_level_values('BodySite')=='Blood']
# Preprocess
X_norm=ICApreprocess(X)
```

### ICA - 

Blood mixtures, $x$ ($M$ by $k$):

$$ \begin{bmatrix} x_1 \\ ... \\ x_M \end{bmatrix}$$

* $M$ blood mixtures.
* Each has $k$ bugs measured. 

Mixing matrix, $A$ ($M$ by $n$):

$$ \begin{bmatrix} a_{11} ... a_{1n} \\ ... \\  a_{M1}  ... a_{Mn} \end{bmatrix} $$

* For each of $M$ blood mixtures, it provides a set of mixing terms.
* Each mixing term is assocaited with one of the $N$ sources.
* Each mixture in $X$ is thus a linear combination given by: $ x_j = a_{j1} s_1 + ... + a_{jn} s_n $

Source matrix, $S$ ($n$ by $k$):

$$ \begin{bmatrix} s_1 \\ ... \\ s_n \end{bmatrix} $$
 
* These are the $n$ independent components of the model.

The inner product of these terms can be written:

* The common set of sources are combined as specified by the mixing matrix to give each.
* Where each mixture, $x_j$, and source, $s_j$,  have the same dimentionality ($k$).

Mixing:

$$ \begin{bmatrix} x_1 \\ ... \\ x_M \end{bmatrix} =  \begin{bmatrix} a_{11} ... a_{1n} \\ ... \\  a_{M1}  ... a_{Mn} \end{bmatrix} \begin{bmatrix} s_1 \\ ... \\ s_n \end{bmatrix} $$

$$ x = AS $$

$$ S = A^{-1}x $$

$$ W = A^{-1} $$

$$ S = Wx $$

Our goal is to find the unmixing matrix, $W$, that allows us to recover the source signals.


#### Compute mean of data pass to ICA for later use - 

```python
# Mean for body sites in DF sampled
mean_dat=pd.DataFrame()
X_nonblood=dat_tp_norm[dat_tp_norm.index.get_level_values('BodySite')!='Blood']
for bs in set(X_nonblood.index.get_level_values('BodySite')):
    bs_dat=dat_tp_norm[dat_tp_norm.index.get_level_values('BodySite')==bs]
    means=bs_dat.mean(axis=0)
    means_norm=means/means.sum()
    mean_dat[bs]=means_norm
```

#### Pre-process data for ICA - 

```python
def ICApreprocess(df):
    ''' Zero out the mean, and made stv one. '''
    # Normalize data by samples with zero mean 
    df_mean_norm=df.subtract(df.mean(axis=1),axis=0)
    # Normalize data by samples with unit variance 
    df_mean_var_norm=df_mean_norm.divide(df_mean_norm.std(axis=1),axis=0)
    return df_mean_var_norm

# Blood data in DF sampled
X=dat_tp_norm[dat_tp_norm.index.get_level_values('BodySite')=='Blood']

# Preprocess
X_norm=ICApreprocess(X)
```

#### Run ICA - 

```python
from pylab import *
from numpy import *
from scipy import stats
from sklearn.decomposition import FastICA

def runICA(X,ica_comp):
    ''' Run ICA with specified number of components''' 

    # ICA
    ica = FastICA(n_components=ica_comp)
    ica.fit(X.T)
    # Write computed source data to DF
    S_est_bugs = ica.transform(X.T)
    ica_sources=pd.DataFrame()
    for i, si in enumerate(S_est_bugs.T):
        ica_sources['Source_%s'%str(i+1)]=si    
    return (ica_sources,ica)

# Compute ICA
ica_sources,ica=runICA(X_norm,7)
ica_sources.index=mean_dat.index
ica_source_coorl=ica_sources.transpose()
```

```python
def computeCoorl(ica_source,df,ica_source_name):
    ''' Compute correlation with HMP or relman data '''
    
    stor=pd.DataFrame(columns=['r2'])
    for bodysite in set(df.index):
        bs=df.loc[bodysite]
        bs=pd.DataFrame(bs)
        cat=bs.merge(ica_source,left_index=True, right_index=True)
        cat.fillna(0,inplace=True)
        slope, intercept, r_value, p_value, std_err = stats.linregress(cat[ica_source_name],cat[bodysite])
        stor.loc[bodysite]=np.round(r_value**2,decimals=2)
    srt=stor.sort('r2',ascending=True)
    top_bs=srt.index[-1] 
    bs_top=pd.DataFrame(df.loc[top_bs])
    cat=bs_top.merge(ica_source,left_index=True, right_index=True, how='outer')
    cat.fillna(0,inplace=True)
    return cat,srt

for ica_source_name in ica_source_coorl.index:

    # ICA sources
    ica_source=ica_source_coorl.loc[ica_source_name]
    ica_source=np.abs(ica_source)
    ica_source=pd.DataFrame(ica_source)
    
    # Relman and HMP mean data
    mean_dat_coorl=mean_dat.transpose()
    hmp_mean_corrl=hmp_dat_norm.transpose()
    
    cat_rel,srt_rel=computeCoorl(ica_source,mean_dat_coorl,ica_source_name)
    cat_hmp,srt_hmp=computeCoorl(ica_source,hmp_mean_corrl,ica_source_name)
    
    del cat_hmp[ica_source_name]
    all_dat=cat_rel.merge(cat_hmp,left_index=True, right_index=True, how='outer')
    all_dat.fillna(0,inplace=True)
    #all_dat.sort(ica_source_name,inplace=True,ascending=False)
    
    # Trim for plotting
    all_dat=all_dat[all_dat[ica_source_name]>0]
    best_rel_site=srt_rel.index[-1]
    best_hmp_site=srt_hmp.index[-1]
    
    # r2 cutoff for plotting
    cutoff=.5
    
    plt.subplot(2,1,1)
    ind=np.arange(len(all_dat.index))
    plt.plot(ind,all_dat[ica_source_name],color='blue',alpha=0.5,label='ICA Source',drawstyle='steps') 
    if srt_rel.loc[best_rel_site,'r2'] > cutoff:
        plt.plot(ind,all_dat[best_rel_site],color='red',alpha=0.5,label='Relman Bodysite %s, $r^2$=%s'%(best_rel_site,srt_rel.loc[best_rel_site,'r2']),drawstyle='steps') 
    if srt_hmp.loc[best_hmp_site,'r2'] > cutoff:
        plt.plot(ind,all_dat[best_hmp_site],color='cyan',alpha=0.5,label='HMP Bodysite %s, $r^2$=%s'%(best_hmp_site,srt_hmp.loc[best_hmp_site,'r2']),drawstyle='steps')
    locs,labels=plt.xticks(ind,all_dat.index)
    plt.setp(labels, rotation=90, fontsize=4)
    plt.yticks(fontsize=5)
    plt.ylabel('Abundance',fontsize=6)
    plt.xlabel('Bug (ranked by abundance in ICA source signal)',fontsize=5)
    plt.xlim([-1,len(all_dat.index)])
    plt.legend(loc='upper left',fontsize=5)
    plt.tick_params(\
    axis='x',         
    which='both',      
    bottom='off',     
    top='off')    
    plt.show()
```

### Covariance - 

Consider that correlation between bugs within a site needs to be high in order to group them into a common source:

https://docs.scipy.org/doc/scipy-0.7.x/reference/spatial.distance.html

http://scikit-learn.org/stable/modules/biclustering.html

```python
print list(set(dat_tp_norm.index.get_level_values('BodySite')))

def makeHeatmap(pruned_DF):
    ''' Heatmap summary of results from regression.'''
    
    data=np.array(pruned_DF,dtype=float)
    sample_Labels=np.array(list(pruned_DF.index))
    bugLabels=np.array(list(pruned_DF.columns))
    
    p=plt.pcolormesh(data,cmap='gist_heat')
    
    locs,labels=plt.yticks(np.arange(sample_Labels.size)+0.5,sample_Labels)
    plt.setp(labels, fontsize=3)
    
    plt.ylim(0,len(sample_Labels))
    plt.tick_params(\
    axis='x',         
    which='both',      
    bottom='off',     
    top='off')         
    # labelbottom='off') 
    
    plt.tick_params(\
    axis='y',          
    which='both',      
    left='off',      
    right='off',        
    labelleft='on')  
    
    ind=np.arange(bugLabels.size)+0.5
    locs,labels=plt.xticks(ind,bugLabels)
    plt.setp(labels, rotation=90, fontsize=3)
    plt.xlim([0,len(bugLabels)]) 
    
bsite='Stool'
bs_dat=dat_tp_norm[dat_tp_norm.index.get_level_values('BodySite')==bsite]
bs_dat_t=bs_dat.T
bug_sum=bs_dat_t.sum(axis=1)
ix=bug_sum[bug_sum>0]
bs_dat_t=bs_dat_t.loc[ix.index]

# Store distance
from scipy import *
stor_corr=pd.DataFrame(scipy.spatial.distance.squareform(scipy.spatial.distance.pdist(bs_dat_t, metric="correlation")), \
                       index=bs_dat_t.index,columns=bs_dat_t.index)
makeHeatmap(stor_corr)
plt.colorbar()
plt.show()

from sklearn.cluster.bicluster import SpectralBiclustering
model = SpectralBiclustering(n_clusters=5, method='log',
                             random_state=0)
model.fit(stor_corr)
fit_data = stor_corr.ix[np.argsort(model.row_labels_),:]
fit_data = fit_data.ix[:, np.argsort(model.row_labels_)]

makeHeatmap(fit_data)
plt.colorbar()
plt.show()
```
