```python
from mpl_toolkits.basemap import Basemap
```

Zoom:

* Specify the latitude and longitude of the lower left and upper right corners of the region.
* Use a mercator projection, which supports this method of zooming.


```python
map = Basemap(projection='ortho', lat_0=0, lon_0=-100,
              resolution='l', area_thresh=1000.0)
map.drawcoastlines()
map.drawcountries()
map.fillcontinents(color='coral')
map.drawmapboundary()
map.drawmeridians(np.arange(0, 360, 30))
map.drawparallels(np.arange(-90, 90, 30))
plt.show()
```
